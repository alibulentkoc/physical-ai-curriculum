```python
import numpy as np
v = np.array([0.30, 0.40])
mag = np.linalg.norm(v)
ang = np.degrees(np.arctan2(v[1], v[0]))
print('magnitude:', round(mag,3),'m   direction:', round(ang,1),'deg')
```
