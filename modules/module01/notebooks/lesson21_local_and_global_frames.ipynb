# Lesson 3.5 Notebook — Local and Global Frames
World stays fixed; local frames move. Run **Kernel → Restart & Run All**.

```python
import numpy as np
tomato_world = np.array([2.0, 1.5])     # global: fixed
robot_world  = np.array([1.4, 1.3])     # robot's origin in the world
def in_robot_frame(p_world, robot_origin):
    return p_world - robot_origin        # axes aligned (no rotation yet)
print('world:', tomato_world, ' robot:', in_robot_frame(tomato_world, robot_world))
```

## Drive the robot east by 0.2 m and re-read both frames

```python
robot_world_moved = robot_world + np.array([0.2, 0.0])
print('after moving:')
print('  world reading :', tomato_world, '(unchanged)')
print('  robot reading :', in_robot_frame(tomato_world, robot_world_moved))
```

## Coding Exercise (§8)
Drive the robot in a full loop back to its start; confirm the robot-frame reading returns to its original value and the world reading never changed.

```python
# YOUR CODE HERE
start = in_robot_frame(tomato_world, robot_world)
for step in [[0.2,0],[0,0.2],[-0.2,0],[0,-0.2]]:
    robot_world = robot_world + np.array(step)
end = in_robot_frame(tomato_world, robot_world)
print('robot reading start:', start, ' end:', end, ' returned:', np.allclose(start,end))
```
