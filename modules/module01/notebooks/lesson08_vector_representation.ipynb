```python
import numpy as np
v = np.array([0.30, 0.40])          # 2D: (x_right, y_up)
w = np.array([0.30, 0.40, 0.15])    # 3D: add z_forward
print('2D components:', v, ' x=',v[0],' y=',v[1])
print('3D components:', w)
```

## Coding Exercise (§8)
Components depend on the chosen frame. Re-express v in a frame whose axes are swapped (up, right) instead of (right, up).

```python
# YOUR CODE HERE
v_swapped = v[[1,0]]
print('original (right,up):', v, ' swapped (up,right):', v_swapped)
```
