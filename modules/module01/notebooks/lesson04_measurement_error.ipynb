# Lesson 1.4 Notebook — Measurement Error

Compute absolute/relative error and see that **averaging reduces random error** but not a systematic offset. Run **Kernel → Restart & Run All**.

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(0)

def absolute_error(measured, true): return abs(measured - true)
def relative_error(measured, true): return abs(measured - true)/abs(true)

print('abs error:', absolute_error(1.20, 1.23))            # 0.03
print('rel error %:', round(relative_error(1.20,1.23)*100,2)) # ~2.44
```

## Random noise vs averaging
Simulate noisy distance readings of a tomato truly at 1.23 m. Watch the running average settle toward the truth.

```python
true_dist = 1.23
readings = true_dist + rng.normal(0, 0.02, size=200)   # random error, no bias
running_avg = np.cumsum(readings)/np.arange(1, len(readings)+1)

plt.figure(figsize=(6,3.5))
plt.axhline(true_dist, color='#0f172a', ls='--', label='true 1.23 m')
plt.plot(readings, '.', alpha=0.3, color='#8b5cf6', label='single readings')
plt.plot(running_avg, color='#10b981', lw=2, label='running average')
plt.xlabel('reading #'); plt.ylabel('distance (m)'); plt.legend(); plt.title('Averaging tames random error')
plt.show()
print('average of all readings:', round(running_avg[-1], 3), 'm')
```

## Coding Exercise (§8)
Given three readings of the 1.23 m target, compute each absolute error and their average. In a comment, say whether averaging helps more against systematic or random error.

```python
# YOUR CODE HERE
three = [1.20, 1.25, 1.22]
errs = [absolute_error(r, 1.23) for r in three]
print('abs errors:', [round(e,3) for e in errs], ' mean reading:', round(sum(three)/3,3))
# Averaging helps against RANDOM error (highs and lows cancel), not a systematic offset.
```
