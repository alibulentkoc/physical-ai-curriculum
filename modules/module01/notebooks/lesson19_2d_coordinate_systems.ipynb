Locate points; quadrant from signs. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def quadrant(p):
    x,y=p
    if x>0 and y>0: return 'I'
    if x<0 and y>0: return 'II'
    if x<0 and y<0: return 'III'
    if x>0 and y<0: return 'IV'
    return 'axis'
for p in [(3,2),(-2,1),(-2,-1),(2,-1)]:
    print(p, '-> quadrant', quadrant(p))
```

```python
dock=np.array([0.5,0.5]); tomato=np.array([3.0,2.0])
disp=tomato-dock
print('displacement dock->tomato:', disp, ' distance:', round(np.linalg.norm(disp),3),'m')
```
