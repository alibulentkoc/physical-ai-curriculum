```python
import numpy as np
gripper=np.array([0.10,0.10,0.05]); tomato=np.array([0.40,0.30,0.25])
disp=tomato-gripper
print('displacement:', disp, ' 3D distance:', round(np.linalg.norm(disp),3),'m')
# right-hand rule: x cross y = z
print('x X y =', np.cross([1,0,0],[0,1,0]), '(= +z, right-handed)')
```

## Coding Exercise (§8)
Three tomatoes in 3D; print the nearest reachable one for a 0.5 m reach.

```python
# YOUR CODE HERE
fruits={'A':[0.40,0.30,0.25],'B':[0.20,0.15,0.10],'C':[0.60,0.50,0.40]}
reach=0.5
d={k:np.linalg.norm(np.array(v)-gripper) for k,v in fruits.items()}
reachable={k:v for k,v in d.items() if v<=reach}
print('distances:', {k:round(v,3) for k,v in d.items()})
print('nearest reachable:', min(reachable, key=reachable.get) if reachable else None)
```
