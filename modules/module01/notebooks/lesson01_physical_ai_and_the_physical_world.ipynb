## 1. A position is a few numbers

The greenhouse robot represents the tomato's position relative to the gripper as three numbers: how far **right**, **up**, and **forward**. Below we work in 2D (right, up) to keep it visual.

```python
# Tomato position relative to the gripper, in meters: (right, up)
tomato = np.array([0.3, 0.4])

right, up = tomato
print(f'Tomato is {right} m right and {up} m up of the gripper.')

# Straight-line distance (a gentle preview of 'magnitude', Lesson 2.5)
distance = np.sqrt(right**2 + up**2)
print(f'Straight-line distance: {distance:.2f} m')
```

```python
# Visualize the gripper at the origin and the tomato as a point, with the arrow between them.
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(0, 0, 'ks', markersize=10, label='gripper (origin)')
ax.plot(tomato[0], tomato[1], 'o', color='#e11d48', markersize=12, label='tomato')
ax.annotate('', xy=tomato, xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', lw=2, color='#334155'))
ax.text(tomato[0]/2 - 0.02, tomato[1]/2 + 0.03, f'{distance:.2f} m', color='#334155')
ax.set_xlim(-0.1, 0.6); ax.set_ylim(-0.1, 0.6)
ax.set_aspect('equal')
ax.set_xlabel('right (m)'); ax.set_ylabel('up (m)')
ax.set_title('Where is the tomato, relative to the gripper?')
ax.grid(True, alpha=0.3); ax.legend()
plt.show()
```

## 2. The perception → reasoning → action loop

The function below *narrates* one cycle of the loop for a given tomato position. It does no real perception or control — it shows how one position flows through the three stages, each speaking a different 'language' (pixels → meters → a decision).

```python
def trace_the_loop(tomato_xy, reach=0.7):
    right, up = tomato_xy
    # PERCEIVE: pretend the camera reported the tomato (here, given directly)
    print('PERCEIVE: camera detects a tomato in view.')
    # REASON: where is it, and can we reach it?
    distance = float(np.sqrt(right**2 + up**2))
    reachable = distance <= reach
    print(f'REASON:   position = ({right} m right, {up} m up), distance = {distance:.2f} m')
    print(f'          reachable within {reach} m? {reachable}')
    # ACT: decide what to do
    if reachable:
        print('ACT:      move gripper toward the tomato and grip.')
    else:
        print('ACT:      out of reach -> reposition the base or skip.')
    return distance, reachable

trace_the_loop(tomato);
```
