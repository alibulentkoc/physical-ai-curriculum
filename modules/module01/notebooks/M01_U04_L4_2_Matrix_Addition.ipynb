```python
import numpy as np
A=np.array([[1,2],[0,1]]); B=np.array([[0,1],[3,2]])
print('A + B =\n', A+B)
print('commutative:', np.allclose(A+B, B+A))
```

## Coding Exercise (§8)
Show addition is NOT composition: (A+B)p vs applying A then B (= B@(A@p)).

```python
# YOUR CODE HERE
p=np.array([1.0,1.0])
add=(A+B)@p
seq=B@(A@p)
print('(A+B)p   =', add)
print('B(Ap)    =', seq)
print('addition is NOT composition:', not np.allclose(add,seq))
```
