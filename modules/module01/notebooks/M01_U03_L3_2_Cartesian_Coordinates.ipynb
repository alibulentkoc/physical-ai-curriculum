```python
import numpy as np
# origin at greenhouse corner; +x east, +y north (meters)
tomato = np.array([2.0, 1.5])    # 2.0 east, 1.5 north
grate  = np.array([-0.5, 1.0])   # 0.5 WEST (negative x), 1.0 north
print('tomato:', tomato, ' grate:', grate)
print('order matters: (2,1) != (1,2) ->', not np.array_equal([2,1],[1,2]))
```
