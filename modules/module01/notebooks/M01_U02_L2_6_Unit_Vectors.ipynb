```python
import numpy as np
v = np.array([0.0, 2.0])
u = v / np.linalg.norm(v)        # unit vector
print('v:', v, ' unit:', u, ' |unit| =', round(np.linalg.norm(u),6))
# step 0.05 m along the heading
print('small step:', 0.05*u)
```
