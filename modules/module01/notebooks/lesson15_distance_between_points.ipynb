# Lesson 2.9 Notebook — Distance Between Points
Distance = magnitude of the connecting vector. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def distance(p,q): return np.linalg.norm(np.array(q)-np.array(p))
gripper=[0.1,0.1,0.0]
fruits={'A':[0.4,0.5,0.1],'B':[0.2,0.2,0.0],'C':[0.7,0.3,0.2]}
for k,f in fruits.items():
    print(k, 'distance', round(distance(gripper,f),3),'m')
nearest=min(fruits, key=lambda k: distance(gripper,fruits[k]))
print('nearest fruit:', nearest)
```

## Coding Exercise (§8)
With a 0.6 m reach, list which fruits are reachable from the gripper.

```python
# YOUR CODE HERE
reach=0.6
reachable=[k for k,f in fruits.items() if distance(gripper,f)<=reach]
print('reachable within', reach,'m:', reachable)
```
