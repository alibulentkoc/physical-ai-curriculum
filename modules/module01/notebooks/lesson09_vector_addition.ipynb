# Lesson 2.3 Notebook — Vector Addition
Tip-to-tail: the resultant is the single equivalent move. Run **Kernel → Restart & Run All**.

```python
import numpy as np
import matplotlib.pyplot as plt
a = np.array([0.30, 0.10])   # move 1
b = np.array([0.10, 0.25])   # move 2
r = a + b                     # resultant
print('a + b =', r, ' (order independent:', np.allclose(a+b, b+a), ')')

plt.figure(figsize=(5,4))
plt.quiver(0,0,a[0],a[1],angles='xy',scale_units='xy',scale=1,color='#0ea5e9',label='a')
plt.quiver(a[0],a[1],b[0],b[1],angles='xy',scale_units='xy',scale=1,color='#8b5cf6',label='b')
plt.quiver(0,0,r[0],r[1],angles='xy',scale_units='xy',scale=1,color='#0d9488',label='a+b')
plt.xlim(-0.05,0.5); plt.ylim(-0.05,0.45); plt.gca().set_aspect('equal')
plt.xlabel('right (m)'); plt.ylabel('up (m)'); plt.legend(); plt.title('Tip-to-tail addition')
plt.show()
```
