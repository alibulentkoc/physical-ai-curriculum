# Lesson 4.5 Notebook — Rotation Matrices
Rotation preserves length. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def R(deg):
    a=np.radians(deg); return np.array([[np.cos(a),-np.sin(a)],[np.sin(a),np.cos(a)]])
p=np.array([1.0,0.0])
for d in [0,45,90,180]:
    q=R(d)@p
    print(f'rotate {d:3d} deg: {q.round(3)}  |q|={np.linalg.norm(q):.3f}')
```

## Coding Exercise (§8)
Rotate a set of shape points by 90° and confirm every point keeps its distance from the origin; check R(90)R(90)=R(180).

```python
# YOUR CODE HERE
pts=np.array([[1,0],[0,1],[2,1],[-1,1]]).T
rot=R(90)@pts
import numpy as np
print('distances preserved:', np.allclose(np.linalg.norm(pts,axis=0), np.linalg.norm(rot,axis=0)))
print('R(90)R(90)=R(180):', np.allclose(R(90)@R(90), R(180)))
```
