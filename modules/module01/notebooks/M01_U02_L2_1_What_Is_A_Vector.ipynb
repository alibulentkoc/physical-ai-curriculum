### Lesson 2.1 — What Is a Vector?

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
# vector from gripper(origin) to tomato: (right, up) in meters
v = np.array([0.30, 0.40])
magnitude = np.linalg.norm(v)            # 'how far' (preview of Lesson 2.5)
print('vector:', v, ' magnitude:', round(magnitude,3), 'm')
# reversing direction is a different vector with the same magnitude
print('reversed:', -v, ' same magnitude:', round(np.linalg.norm(-v),3))
```
