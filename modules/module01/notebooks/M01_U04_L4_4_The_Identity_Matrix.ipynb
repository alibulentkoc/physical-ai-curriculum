### Lesson 4.4 — The Identity Matrix

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
I = np.eye(2)
print('I =\n', I)
for p in [np.array([0,0]),np.array([2,1]),np.array([-3,4])]:
    print(p, '-> I @ p =', I @ p)
print('every point unchanged.')
```

## Coding Exercise (§8)
Show I is the neutral element: I @ M == M @ I == M for some M; contrast a non-identity that moves a point.

```python
# YOUR CODE HERE
M = np.array([[2.0,0.0],[0.0,2.0]])
print('I@M == M:', np.allclose(I@M, M), ' M@I == M:', np.allclose(M@I, M))
print('non-identity moves (2,1):', (M @ np.array([2,1])))
```

## Check your work

```python
assert np.allclose(I @ np.array([2,1]), [2,1])
assert np.allclose(I@M, M) and np.allclose(M@I, M)
print('All checks passed.')
```
