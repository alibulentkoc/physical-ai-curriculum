# Lesson 2.4 Notebook — Vector Subtraction
b − a is the move from a to b. Run **Kernel → Restart & Run All**.

```python
import numpy as np
gripper = np.array([0.20, 0.10])   # a
tomato  = np.array([0.55, 0.40])   # b
move = tomato - gripper             # b - a
print('move the arm by:', move, ' distance:', round(np.linalg.norm(move),3),'m')
print('reverse (b->a):', gripper - tomato)
```

## Coding Exercise (§8)
Compute the move from the tomato back to a basket at (0.0, 0.0). Confirm move(a→b) = −move(b→a).

```python
# YOUR CODE HERE
basket = np.array([0.0, 0.0])
to_basket = basket - tomato
print('tomato -> basket:', to_basket, ' opposite of basket->tomato:', np.allclose(to_basket, -(tomato-basket)))
```
