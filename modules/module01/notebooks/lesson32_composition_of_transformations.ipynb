# Lesson 4.8 Notebook â€” Composition of Transformations
Order matters: scaleâ†’rotate â‰  rotateâ†’scale. Run **Kernel â†’ Restart & Run All**.

```python
import numpy as np
def R(deg):
    a=np.radians(deg); return np.array([[np.cos(a),-np.sin(a)],[np.sin(a),np.cos(a)]])
S=np.array([[2.0,0.0],[0.0,1.0]])    # stretch x only
p=np.array([1.0,0.0])
scale_then_rotate = R(90)@S          # S first (on the right)
rotate_then_scale = S@R(90)
print('scale then rotate, p ->', np.round(scale_then_rotate@p,3))
print('rotate then scale, p ->', np.round(rotate_then_scale@p,3))
print('different outcome:', not np.allclose(scale_then_rotate@p, rotate_then_scale@p))
```

## Conceptual translation in the chain
Translation isn't a 2x2 matrix (Module 2 formalizes it). Here we add it as an action: rotate a point, then translate it, vs translate then rotate.

```python
t=np.array([1.0,0.0])    # a move
p=np.array([1.0,0.0])
rotate_then_translate = R(90)@p + t
translate_then_rotate = R(90)@(p + t)
print('rotate then translate:', np.round(rotate_then_translate,3))
print('translate then rotate:', np.round(translate_then_rotate,3))
print('order matters with translation too:', not np.allclose(rotate_then_translate, translate_then_rotate))
```
