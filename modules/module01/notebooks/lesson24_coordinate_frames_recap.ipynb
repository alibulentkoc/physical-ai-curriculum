# Lesson 3.8 Notebook — Coordinate Frames Capstone (Unit 3 Recap)
One fixed tomato, three frames, one conversion. Run **Kernel → Restart & Run All**.

```python
import numpy as np
# one fixed tomato, described three ways (axes aligned for clarity)
world  = np.array([2.0, 1.5])
robot_origin_in_world  = np.array([1.4, 1.3])
camera_origin_in_world = np.array([1.7, 1.6])
robot  = world - robot_origin_in_world
camera = world - camera_origin_in_world
print('world :', world)
print('robot :', robot)
print('camera:', camera)
print('Same tomato. Three observers. It never moved.')
```

## Coding Exercise (§8)
Convert the tomato from the robot frame to the camera frame using the offset between their origins, and confirm it matches the camera reading above.

```python
# YOUR CODE HERE
offset_robot_to_camera = camera_origin_in_world - robot_origin_in_world
camera_from_robot = robot - offset_robot_to_camera
print('camera (via robot):', camera_from_robot, ' matches direct:', np.allclose(camera_from_robot, camera))
```

## Check your work

```python
assert np.allclose(robot, [0.6,0.2])
assert np.allclose(camera, [0.3,-0.1])
assert np.allclose(camera_from_robot, camera)   # transformation consistency
print('All checks passed.')
```
