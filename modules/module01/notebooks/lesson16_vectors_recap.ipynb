```python
import numpy as np
gripper = np.array([0.10, 0.10, 0.00])   # 2.1 position
tomato  = np.array([0.45, 0.40, 0.15])   # 2.1 position

move      = tomato - gripper             # 2.4 displacement
distance  = np.linalg.norm(move)         # 2.5/2.9 magnitude / distance
heading   = move / distance              # 2.6 unit vector (direction)
print('move (b-a):', move.round(2))
print('distance:', round(distance,3), 'm   heading (unit):', heading.round(2))

reach = 0.7
print('reachable?' , distance <= reach)   # 2.9 reachability
```

```python
# 2.7 dot product: is the gripper (facing +x) aligned with the fruit direction?
facing = np.array([1.0, 0.0, 0.0])
alignment = np.dot(facing, heading)
print('alignment (dot):', round(alignment,2), '-> facing' if alignment>0 else '-> not facing')

# 2.8 cross product: surface normal from two edge vectors of a leaf/face
e1 = np.array([0.2, 0.0, 0.0]); e2 = np.array([0.0, 0.2, 0.1])
normal = np.cross(e1, e2)
print('surface normal (a x b):', normal.round(3), ' area:', round(np.linalg.norm(normal),3))
```
