# Lesson 2.7 Notebook — Dot Product
Alignment: positive when aligned, zero when perpendicular, negative when opposed. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def angle_between(a,b):
    c = np.dot(a,b)/(np.linalg.norm(a)*np.linalg.norm(b))
    return np.degrees(np.arccos(np.clip(c,-1,1)))
gripper_dir = np.array([1.0, 0.0])
for name,target in [('aligned',[1,0.1]),('perpendicular',[0,1]),('opposed',[-1,0.1])]:
    t=np.array(target,float)
    print(f'{name:13s} dot={np.dot(gripper_dir,t):+.2f}  angle={angle_between(gripper_dir,t):5.1f} deg')
```
