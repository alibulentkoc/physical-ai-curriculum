```python
import numpy as np
def R(deg):
    a=np.radians(deg); return np.array([[np.cos(a),-np.sin(a)],[np.sin(a),np.cos(a)]])
I=np.eye(2)
S=np.array([[2,0],[0,0.5]])
F=np.array([[1,0],[0,-1]])   # reflect across x
p=np.array([1.0,1.0])
print('identity  :', I@p)
print('rotate 90 :', np.round(R(90)@p,3))
print('scale     :', S@p)
print('reflect   :', F@p)
```

## Coding Exercise (§8)
Compose rotate and scale in both orders on one point and confirm the results differ.

```python
# YOUR CODE HERE
import numpy as np
p=np.array([1.0,0.0])
a=(R(90)@S)@p   # scale then rotate
b=(S@R(90))@p   # rotate then scale
print('scale->rotate:', np.round(a,3), ' rotate->scale:', np.round(b,3), ' differ:', not np.allclose(a,b))
```
