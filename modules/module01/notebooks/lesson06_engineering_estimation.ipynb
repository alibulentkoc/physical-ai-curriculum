# Lesson 1.6 Notebook — Engineering Estimation

Order-of-magnitude reasoning to sanity-check a claim before any detailed model. Run **Kernel → Restart & Run All**.

```python
import math

seconds_per_pick = 10      # rough guess
seconds_per_hour = 3600
picks_per_hour = seconds_per_hour / seconds_per_pick
print(f'~{picks_per_hour:.0f} picks/hour per arm  (order 10^{int(round(math.log10(picks_per_hour)))})')
```

## Test a claim
A proposal claims 5000 picks/hour on one arm. Compare orders of magnitude.

```python
claim = 5000
ratio = claim / picks_per_hour
print(f'claim is ~{ratio:.0f}x our estimate -> order 10^{int(round(math.log10(claim)))} vs 10^{int(round(math.log10(picks_per_hour)))}')
print('Implausible for a single arm.' if ratio > 3 else 'Plausible.')
```

## Coding Exercise (§8)
Change `seconds_per_pick` to a faster and a slower guess; print the order of magnitude each time and whether 5000/hour becomes plausible.

```python
# YOUR CODE HERE
for spp in (3, 10, 30):
    pph = 3600/spp
    print(f'{spp:2d} s/pick -> {pph:.0f} picks/hr (order 10^{int(round(math.log10(pph)))})')
```
