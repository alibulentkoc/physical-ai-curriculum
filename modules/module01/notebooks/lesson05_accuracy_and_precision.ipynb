# Lesson 1.5 Notebook — Accuracy and Precision

Quantify **accuracy** (bias = how far the average is from true) and **precision** (spread) from sample readings, and visualize the dartboard idea. Run **Kernel → Restart & Run All**.

```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng(0)

readings = [1.78, 1.80, 1.79, 1.81, 1.80]
true_value = 2.00
avg = sum(readings)/len(readings)
bias = avg - true_value          # accuracy indicator
spread = max(readings)-min(readings)  # crude precision indicator
print(f'avg={avg:.3f}  bias={bias:.3f} (accuracy)  spread={spread:.3f} (precision)')
print('-> precise (tight) but not accurate (biased): fix with a +%.2f m calibration offset' % (-bias))
```

## Dartboard view
Two clusters: precise-but-biased vs accurate-but-scattered, around the bullseye (true).

```python
fig, ax = plt.subplots(1,2, figsize=(7,3.6))
for a in ax:
    for r in (1.0,0.6,0.25):
        a.add_patch(plt.Circle((0,0), r, fill=False, color='#cbd5e1'))
    a.set_xlim(-1.2,1.2); a.set_ylim(-1.2,1.2); a.set_aspect('equal'); a.axis('off')
# precise but biased
p = rng.normal([0.6,0.5],0.05,size=(8,2)); ax[0].plot(p[:,0],p[:,1],'o',color='#0f172a'); ax[0].set_title('precise, not accurate')
# accurate but scattered
q = rng.normal([0,0],0.45,size=(8,2)); ax[1].plot(q[:,0],q[:,1],'o',color='#0f172a'); ax[1].set_title('accurate, not precise')
plt.show()
```
