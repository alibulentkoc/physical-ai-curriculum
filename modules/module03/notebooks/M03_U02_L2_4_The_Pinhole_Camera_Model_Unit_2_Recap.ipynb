```python
import numpy as np
def project(P,f): X,Y,Z=P; return np.array([f*X/Z, f*Y/Z])
P=np.array([0.06,0.0,0.3])
print('ray sharing:', np.allclose(project(P,800), project(P*2,800)))
print('scales with f:', np.allclose(project(P,1600), 2*project(P,800)))
print('scales with 1/Z:', np.allclose(project(np.array([0.06,0,0.6]),800), 0.5*project(P,800)))
```
