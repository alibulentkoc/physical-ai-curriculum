> **OpenCV note:** This is Module 3's first use of OpenCV (`cv2`). The notebook also includes a pure-NumPy implementation of the identical formula, so it runs even if OpenCV is unavailable.

```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def project_numpy(pts, R, t, K):
    pts=np.asarray(pts,float)
    Pc=(R@pts.T).T + t
    p=(K@Pc.T).T
    return p[:,:2]/p[:,2:3]
# OpenCV if available, else fall back to the same NumPy formula
try:
    import cv2
    HAVE_CV2=True
    def project_cv(pts,rvec,tvec,K):
        out,_=cv2.projectPoints(np.asarray(pts,float), np.asarray(rvec,float), np.asarray(tvec,float), K, np.zeros(5))
        return out.reshape(-1,2)
except Exception:
    HAVE_CV2=False
print('OpenCV available:', HAVE_CV2)
```

## Coding Exercise (§8)
Project the same points by hand and via OpenCV (or the NumPy fallback); assert they agree.

```python
# YOUR CODE HERE
import numpy as np
pts=[[0.06,-0.03,0.3],[0.0,0.0,0.5]]
R=np.eye(3); t=np.zeros(3); rvec=np.zeros(3); tvec=np.zeros(3)
by_hand=project_numpy(pts,R,t,K)
print('by hand:\n', np.round(by_hand,2))
if HAVE_CV2:
    via_cv=project_cv(pts,rvec,tvec,K)
    print('opencv:\n', np.round(via_cv,2))
    print('match:', np.allclose(by_hand,via_cv,atol=1e-6))
```
