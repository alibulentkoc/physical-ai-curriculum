```python
import numpy as np
import matplotlib; matplotlib.use('Agg'); import matplotlib.pyplot as plt
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def transform(T,P): return (T@np.append(np.asarray(P,float),1.0))[:3]
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
T_wc=np.eye(4); T_wc[:3,3]=[1.0,0.5,0.1]
Pc=np.array([0.06,-0.03,0.3]); Pw=transform(T_wc,Pc)
def verify(Pw,pixel,K,T_wc,Pc,workspace=1.5):
    Pc_back=transform(np.linalg.inv(T_wc),Pw)
    reproj=np.linalg.norm(project_K(Pc_back,K)-np.array(pixel))
    t=T_wc[:3,3]; dist_ok=np.isclose(np.linalg.norm(Pc),np.linalg.norm(Pw-t))
    return {'reproj_px':float(reproj),'reproj_ok':reproj<1.0,'distance_ok':bool(dist_ok),'workspace_ok':bool(np.linalg.norm(Pw)<workspace)}
print(verify(Pw,(480,160),K,T_wc,Pc))
```

## Coding Exercise (§8)
All checks pass for canonical; an injected depth error trips the distance check; faux-3D plot.

```python
# YOUR CODE HERE
import numpy as np
r=verify(Pw,(480,160),K,T_wc,Pc)
assert r['reproj_ok'] and r['distance_ok'] and r['workspace_ok']
# inject depth error: P_c computed with wrong Z but P_w from right Z mismatch -> distance fails
Pc_bad=np.array([0.12,-0.06,0.6])  # would be Z=0.6
r2=verify(Pw,(480,160),K,T_wc,Pc_bad)
assert not r2['distance_ok']
# visualization (headless)
fig=plt.figure(); ax=fig.add_subplot(111,projection='3d')
ax.scatter(*Pc,c='r'); ax.plot([0,Pc[0]],[0,Pc[1]],[0,Pc[2]],c='orange'); ax.set_title('camera/ray/fruit')
fig.savefig('/tmp/verify.png'); print('checks pass; depth-error trips distance; plot saved')
```
