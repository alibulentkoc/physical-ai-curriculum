### Lesson 2.1 — The Pinhole Idea

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def pinhole(P,f): X,Y,Z=P; return np.array([f*X/Z, f*Y/Z])
P=np.array([0.06,0.0,0.3])
for f in [400,800,1200]:
    print('f=%d -> image %s'%(f, np.round(pinhole(P,f),1)))
```

## Coding Exercise (§8)
Show points on the same ray share an image point, and image scales with f.

```python
# YOUR CODE HERE
import numpy as np
a=pinhole(np.array([0.06,0,0.3]),800)
b=pinhole(np.array([0.12,0,0.6]),800)
print('same ray same image:', np.allclose(a,b))
print('double f doubles image:', np.allclose(pinhole(P,1600), 2*pinhole(P,800)))
```
