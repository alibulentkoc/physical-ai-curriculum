```python
import numpy as np
f=800.0
def project(P): X,Y,Z=P; return np.array([f*X/Z, f*Y/Z])
def ortho(P): X,Y,Z=P; return np.array([f*X, f*Y])  # depth-independent (for contrast)
P=np.array([0.06,-0.03,0.3])
print('perspective:', np.round(project(P),1))
print('move to Z=0.6:', np.round(project(np.array([0.06,-0.03,0.6])),1), '(halved)')
```

## Coding Exercise (§8)
Project a row of equally spaced receding points; show image spacing is NOT equal (convergence).

```python
# YOUR CODE HERE
import numpy as np
xs=[project(np.array([0.1,0,Z]))[0] for Z in [0.4,0.6,0.8,1.0]]
gaps=np.diff(xs)
print('image x positions:', np.round(xs,1))
print('gaps shrink (convergence):', np.all(np.abs(np.diff(gaps))>1e-9))
```
