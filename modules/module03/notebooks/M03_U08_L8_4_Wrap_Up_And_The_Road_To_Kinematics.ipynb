```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def transform(T,P): return (T@np.append(np.asarray(P,float),1.0))[:3]
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
def deproject(u,v,Z,K): return np.array([(u-K[0,2])*Z/K[0,0],(v-K[1,2])*Z/K[1,1],Z])
# T_world<-arm was GIVEN by Module 3; Module 4 (forward kinematics) will COMPUTE it.
T_wc=np.eye(4); T_wc[:3,3]=[1.0,0.5,0.1]
Pc=deproject(480,160,0.3,K); Pw=transform(T_wc,Pc)
checks={'reproj':np.linalg.norm(project_K(transform(np.linalg.inv(T_wc),Pw),K)-[480,160])<1.0,
        'distance':np.isclose(np.linalg.norm(Pc),np.linalg.norm(Pw-T_wc[:3,3])),
        'workspace':np.linalg.norm(Pw)<1.5}
print('P_w:',np.round(Pw,3),' checks:',checks)
```
