```python
import numpy as np
# An image is a 2D array of pixels. Place a bright 'fruit' blob and find its pixel.
H,W=48,64
img=np.zeros((H,W))
cy,cx=24,32
yy,xx=np.mgrid[0:H,0:W]
img=np.exp(-(((xx-cx)**2+(yy-cy)**2)/40.0))  # gaussian blob
# centroid (pixel location of the fruit)
total=img.sum(); u=(xx*img).sum()/total; v=(yy*img).sum()/total
print('fruit pixel (u,v) = (%.1f, %.1f)'%(u,v))
print('this is a 2D pixel, NOT a 3D position')
```
