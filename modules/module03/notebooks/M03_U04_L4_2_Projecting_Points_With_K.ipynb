### Lesson 4.2 — Projecting Points with K

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]]); W,H=640,480
P=np.array([[0,0,0.5],[0.2,0,0.4],[0,0,-0.3]]).T  # 3xN camera-frame points
def project_batch(P,K,W,H):
    in_front = P[2]>0
    Z=np.where(in_front, P[2], 1.0)
    p=K@P; uv=(p[:2]/Z)
    in_frame = in_front & (uv[0]>=0)&(uv[0]<W)&(uv[1]>=0)&(uv[1]<H)
    return uv, in_front, in_frame
uv,inf,inf2=project_batch(P,K,W,H)
print('pixels:\n', np.round(uv.T,1))
print('in front:', inf, ' in frame:', inf2)
```

## Coding Exercise (§8)
Reproduce the worked-example classification: A in-frame, B out-of-frame, C behind.

```python
# YOUR CODE HERE
import numpy as np
labels=['A','B','C']
for i,name in enumerate(labels):
    state = 'behind' if not inf[i] else ('in-frame' if inf2[i] else 'out-of-frame')
    print(name, np.round(uv[:,i],1) if inf[i] else '(invalid)', '->', state)
```
