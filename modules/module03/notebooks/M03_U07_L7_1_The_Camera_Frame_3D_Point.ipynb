### Lesson 7.1 — The Camera-Frame 3D Point

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def transform(T,P):
    Ph=np.append(np.asarray(P,float),1.0); return (T@Ph)[:3]
Pc=np.array([0.06,-0.03,0.3])
print('identity extrinsics -> P_w == P_c:', transform(np.eye(4),Pc))
print('translation (1,0.5,0):', transform(se3(np.eye(3),[1.0,0.5,0.0]),Pc))
```

## Coding Exercise (§8)
Identity ⇒ P_w=P_c; translation-only shifts the point.

```python
# YOUR CODE HERE
import numpy as np
assert np.allclose(transform(np.eye(4),Pc),Pc)
assert np.allclose(transform(se3(np.eye(3),[1.0,0.5,0.0]),Pc),[1.06,0.47,0.3])
print('camera-frame point transforms as expected.')
```
