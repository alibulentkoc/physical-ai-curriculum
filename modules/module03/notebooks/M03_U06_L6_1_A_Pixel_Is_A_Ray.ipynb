## Unit 6 — Back-Projection: Pixels to 3D
### Lesson 6.1 — A Pixel Is a Ray

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def backproject_ray(u,v,K):
    xn=(u-K[0,2])/K[0,0]; yn=(v-K[1,2])/K[1,1]
    return np.array([xn,yn,1.0])
d=backproject_ray(480,160,K); print('direction:', d)
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
for lam in [0.3,0.5,0.8]:
    print('lambda',lam,'-> point',np.round(lam*d,3),'-> pixel',np.round(project_K(lam*d,K),1))
```

## Coding Exercise (§8)
Show several depths along the ray all project back to the same pixel; principal point → (0,0,1).

```python
# YOUR CODE HERE
import numpy as np
for lam in [0.3,0.5,0.8]:
    assert np.allclose(project_K(lam*d,K),[480,160])
assert np.allclose(backproject_ray(320,240,K),[0,0,1])
print('all depths reproject to (480,160); principal point -> (0,0,1)')
```
