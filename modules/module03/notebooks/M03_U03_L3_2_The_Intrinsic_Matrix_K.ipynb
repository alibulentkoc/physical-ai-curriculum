### Lesson 3.2 — The Intrinsic Matrix K

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def make_K(fx,fy,cx,cy): return np.array([[fx,0,cx],[0,fy,cy],[0,0,1.0]])
def project_K(P,K):
    p=K@np.asarray(P,float); return p[:2]/p[2]
K=make_K(800,800,320,240)
print('K=\n',K)
print('point (0.06,-0.03,0.3) -> pixel', np.round(project_K([0.06,-0.03,0.3],K),1))
print('optical-axis point (0,0,0.5) -> pixel', np.round(project_K([0,0,0.5],K),1))
```
