```python
import numpy as np
def fov_deg(w,f): return 2*np.degrees(np.arctan(w/f))
w=12.0  # sensor half-width (mm)
for f in [12,24,36]:
    print('f=%d mm -> FOV %.1f deg'%(f, fov_deg(w,f)))
```

## Coding Exercise (§8)
Confirm larger f → smaller FOV, and magnification (image displacement) grows with f.

```python
# YOUR CODE HERE
import numpy as np
print('FOV shrinks with f:', fov_deg(12,36) < fov_deg(12,12))
disp=lambda f: f*0.06/0.3
print('magnification grows with f:', disp(36) > disp(12))
```
