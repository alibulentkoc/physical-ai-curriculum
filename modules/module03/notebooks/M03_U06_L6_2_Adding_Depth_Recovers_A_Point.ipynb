### Lesson 6.2 — Adding Depth Recovers a Point

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def deproject(u,v,Z,K):
    X=(u-K[0,2])*Z/K[0,0]; Y=(v-K[1,2])*Z/K[1,1]; return np.array([X,Y,Z])
print('Z=0.3:', deproject(480,160,0.3,K))
print('Z=0.6:', deproject(480,160,0.6,K))
```
