```python
import numpy as np
import matplotlib; matplotlib.use('Agg'); import matplotlib.pyplot as plt
# straight grid of normalized points
g=np.linspace(-0.5,0.5,11)
X,Y=np.meshgrid(g,g)
def radial(X,Y,k1):
    r2=X**2+Y**2; f=1+k1*r2; return X*f, Y*f
Xb,Yb=radial(X,Y, 0.8)   # barrel-ish (k1>0)
Xp,Yp=radial(X,Y,-0.8)   # pincushion-ish (k1<0)
print('center unchanged (barrel):', np.allclose([Xb[5,5],Yb[5,5]],[0,0]))
print('edge magnitude grows with k1>0:', abs(Xb[0,0])>abs(X[0,0]))
```
