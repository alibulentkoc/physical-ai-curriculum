```python
import numpy as np
f=800.0
def project(P):
    X,Y,Z=P; return np.array([f*X/Z, f*Y/Z])
# Forward is many-to-one: points on the same ray share a pixel
ray_dir=np.array([0.2,0.0,1.0])
for Z in [0.3,0.6,1.2]:
    P=ray_dir*Z
    print('Z=%.1f -> pixel %s'%(Z, np.round(project(P),1)))
```

## Coding Exercise (§8)
Show recovering the 3D point from a pixel needs a chosen depth.

```python
# YOUR CODE HERE
import numpy as np
pix=project(ray_dir*0.3)  # some pixel
# back-project: a pixel gives a direction; multiply by chosen depth Z to get the point
def backproject(pix, Z): return np.array([pix[0]*Z/f, pix[1]*Z/f, Z])
print('depth 0.3 ->', np.round(backproject(pix,0.3),3))
print('depth 0.6 ->', np.round(backproject(pix,0.6),3), '(same pixel, different point)')
```
