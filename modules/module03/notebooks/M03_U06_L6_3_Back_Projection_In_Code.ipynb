```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def deproject_many(pixels, depths, K):
    uv=np.asarray(pixels,float); Z=np.asarray(depths,float)
    valid=Z>0
    X=(uv[:,0]-K[0,2])*Z/K[0,0]; Y=(uv[:,1]-K[1,2])*Z/K[1,1]
    P=np.stack([X,Y,Z],axis=1)
    return P[valid], valid
px=[[480,160],[320,240],[100,100]]; zs=[0.3,0.5,0.0]
P,valid=deproject_many(px,zs,K)
print('valid mask:',valid); print('points:\n',np.round(P,3))
```

## Coding Exercise (§8)
Round-trip: project the deprojected points back to pixels; invalid depth filtered out.

```python
# YOUR CODE HERE
import numpy as np
def project_many(P,K): p=(K@P.T).T; return p[:,:2]/p[:,2:3]
back=project_many(P,K)
print('reprojected:\n',np.round(back,1))
assert np.allclose(back[0],[480,160]) and np.allclose(back[1],[320,240])
assert valid.tolist()==[True,True,False]
```
