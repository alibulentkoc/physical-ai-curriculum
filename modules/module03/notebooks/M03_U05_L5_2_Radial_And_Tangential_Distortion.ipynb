### Lesson 5.2 — Radial and Tangential Distortion

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def distort(xn,yn,k1=0,k2=0,k3=0,p1=0,p2=0):
    r2=xn*xn+yn*yn
    rad=1+k1*r2+k2*r2**2+k3*r2**3
    xr,yr=xn*rad, yn*rad
    xt=2*p1*xn*yn+p2*(r2+2*xn*xn)
    yt=p1*(r2+2*yn*yn)+2*p2*xn*yn
    return xr+xt, yr+yt
print('identity (all zero):', distort(0.4,0.3))
print('k1=-0.2:', tuple(round(v,4) for v in distort(0.4,0.3,k1=-0.2)))
```
