### Lesson 1.3 — What Projection Keeps and Discards

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
f=800.0
def apparent_size(true_size, Z): return f*true_size/Z   # ~ 1/Z
for Z in [0.3,0.6,1.2]:
    print('Z=%.1f m -> apparent size %.1f px'%(Z, apparent_size(0.06,Z)))
```

## Coding Exercise (§8)
Confirm apparent size ∝ 1/Z; find two (size, depth) pairs with equal apparent size.

```python
# YOUR CODE HERE
import numpy as np
a1=apparent_size(0.06,0.6)
a2=apparent_size(0.12,1.2)
print('6cm@0.6m: %.1f px ; 12cm@1.2m: %.1f px ; equal: %s'%(a1,a2,np.isclose(a1,a2)))
```
