### Lesson 7.3 — Estimating the Fruit's World Position

```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def transform(T,P): return (T@np.append(np.asarray(P,float),1.0))[:3]
def deproject(u,v,Z,K): return np.array([(u-K[0,2])*Z/K[0,0],(v-K[1,2])*Z/K[1,1],Z])
def pixel_to_world(u,v,Z,K,T_wc):
    Pc=deproject(u,v,Z,K); return transform(T_wc,Pc), Pc
T_wc=se3(np.eye(3),[0,0,0.1]); T_wc=se3(np.eye(3),[1.0,0.5,0.0])@T_wc
Pw,Pc=pixel_to_world(480,160,0.3,K,T_wc)
print('P_c:',Pc,'  P_w:',np.round(Pw,3))
```

## Coding Exercise (§8)
Verify P_w; distance preservation; re-projection round-trip.

```python
# YOUR CODE HERE
import numpy as np
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
assert np.allclose(Pw,[1.06,0.47,0.4])
# distance preserved: ||P_c|| == ||P_w - t_wc||
t=T_wc[:3,3]; assert np.allclose(np.linalg.norm(Pc),np.linalg.norm(Pw-t))
# re-projection: P_w -> camera -> pixel
Pc_back=transform(np.linalg.inv(T_wc),Pw)
assert np.allclose(project_K(Pc_back,K),[480,160])
print('P_w verified; distance preserved; re-projection OK.')
```
