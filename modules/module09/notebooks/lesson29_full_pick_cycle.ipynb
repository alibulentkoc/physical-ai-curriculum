# Lesson 8.1 — The Full Pick Cycle: Six Stages as One Loop

One pick runs all six stages — Perceive, Understand, Plan, Execute, Track, Recover — as a single self-healing loop.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, Fruit, model_perception,
    understand, to_configuration, recover, harvest_row)
W = GreenhouseWorld.demo_row(n=6, seed=1)
DETS = model_perception(W, rng=np.random.default_rng(7))
TGT = understand(DETS, W)[1]            # nearest ripe reachable target
kick = lambda t, j: 60.0 if (j == 0 and t > 0.3) else 0.0
checks = []
# Healthy pick: forward path runs once, Track passes, nothing to recover
h = recover(W, W.q.copy(), target=TGT, disturbance=lambda a: None)
print('HEALTHY   -> success=%s recovered=%s n_attempts=%d' % (h['success'], h['recovered'], h['n_attempts']))
checks.append(h['success'] and h['recovered'] is False and h['n_attempts'] == 1)
```

### A transient disturbance: the same loop recovers (Execute runs twice)

```python
t = recover(W, W.q.copy(), target=TGT, disturbance=lambda a: (kick if a == 0 else None))
print('TRANSIENT -> success=%s recovered=%s n_attempts=%d | fault then retry-execute'
      % (t['success'], t['recovered'], t['n_attempts']))
checks.append(t['success'] and t['recovered'] is True and t['n_attempts'] == 2)
# only the recovery bookkeeping differs between the two picks
checks.append(h['success'] == t['success'] == True)
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
