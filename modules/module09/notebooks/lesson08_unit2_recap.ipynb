# Lesson 2.4 — Unit 2 Recap (Consolidated Seam Check)

One run of the whole Perceive → Understand seam: perceive with a duplicate and noise, clean + select, then occlude the winner and re-select.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, model_perception, understand)
checks = []
world = GreenhouseWorld.demo_row(n=6, seed=5)
for i,f in enumerate(world.fruit):   # ensure several ripe+reachable
    f.ripe = (i % 4 != 0)
    r = 0.6 + 0.12*i; f.x, f.y = r, 0.3
```

```python
dets = model_perception(world, rng=np.random.default_rng(5), noise=0.01, duplicate=['F2'])
targets, current = understand(dets, world)
print('raw detections:', len(dets), '| world-state targets:', len(targets),
      '| committed:', current['id'] if current else None)
checks.append(len(targets) == len(world.fruit))            # duplicate merged
checks.append(current is None or (current['ripe'] and current['reachable']))
```

### Occlude the winner and re-select

```python
if current:
    occ = model_perception(world, rng=np.random.default_rng(5), occlude=[current['id']])
    _, current2 = understand(occ, world)
    print('occluded', current['id'], '-> new committed:', current2['id'] if current2 else None)
    checks.append(current2 is None or current2['id'] != current['id'])
else:
    checks.append(True)
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
