```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, Fruit, model_perception,
    understand, recover, RECOVERY_POLICY)
W = GreenhouseWorld.demo_row(n=6, seed=1)
ALL_IDS = [f.fid for f in W.fruit]
TGT = understand(model_perception(W, rng=np.random.default_rng(7)), W)[1]
def kick(t, j): return 60.0 if (j == 0 and t > 0.3) else 0.0
checks = []
TGT = understand(model_perception(W, rng=np.random.default_rng(7)), W)[1]
outcomes = {}
outcomes['transient']     = recover(W, W.q.copy(), disturbance=lambda a: kick if a==0 else None)
outcomes['persistent']    = recover(W, W.q.copy(), disturbance=lambda a: kick, max_attempts=3)
outcomes['deterministic'] = recover(W, W.q.copy(), obstacle=(TGT['xy'], 0.25))
outcomes['healthy']       = recover(W, W.q.copy())
for name, r in outcomes.items():
    tag = 'recovered' if r['recovered'] else ('escalated:'+r.get('escalation','') if r['escalated'] else 'first-try')
    print('%-14s -> success=%-5s n=%d  %s' % (name, str(r['success']), r['n_attempts'], tag))
checks.append(outcomes['transient']['recovered'] is True)
checks.append(outcomes['persistent']['escalation'] == 'retry-budget-exhausted')
checks.append(outcomes['deterministic']['escalation'] == 'skip-target')
checks.append(outcomes['healthy']['success'] and outcomes['healthy']['n_attempts'] == 1)
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('UNIT 7 COMPLETE: Recover coordinates, routes, and terminates on the real layers.')
print('All checks passed.')
```
