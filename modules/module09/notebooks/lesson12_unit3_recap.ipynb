# Lesson 3.4 — Unit 3 Recap (Consolidated Chain)

One run of the whole Understand → Plan seam: perceive, understand, convert to configuration, plan, and confirm the endpoint reaches the target. Then confirm an out-of-reach target stops at IK.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, model_perception, understand,
    to_configuration, plan_reference, fk_xy, P2, T2, REACH_MAX)
checks = []
world = GreenhouseWorld.demo_row(n=6, seed=2)
dets = model_perception(world, rng=np.random.default_rng(2))
_, tgt = understand(dets, world)
print('target:', tgt['id'], np.round(tgt['xy'],3))
q_goal = to_configuration(tgt)
layer = plan_reference(world.q.copy(), q_goal, rng=np.random.default_rng(0))
end = fk_xy(P2, T2, layer['reference'](layer['duration'])[0])
print('validated:', layer['validated'], '| endpoint:', np.round(end,3),
      '| reaches target:', bool(np.linalg.norm(end-tgt['xy'])<1e-3))
checks.append(tgt['ripe'] and tgt['reachable'])
checks.append(np.linalg.norm(fk_xy(P2,T2,q_goal)-tgt['xy']) < 1e-9)
checks.append(layer['validated'] and np.linalg.norm(end-tgt['xy']) < 1e-3)
```

```python
# out-of-reach -> chain stops at IK with None
q_none = to_configuration({'xy': np.array([REACH_MAX+0.5, 0.0])})
print('out-of-reach -> to_configuration:', q_none)
checks.append(q_none is None)
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
