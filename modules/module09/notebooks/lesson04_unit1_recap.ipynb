# Lesson 1.4 — Unit 1 Recap (Consolidated Check)

One cell to prove the Unit 1 lens is installed and the foundation runs: print the spine, run Perceive+Understand, assert the canonical order and a reachable target.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, model_perception,
    understand, LAYER_REGISTRY, STAGES)
checks = []
for e in LAYER_REGISTRY:
    print(f"{e['stage']:11s} -> {e['module']}")
checks.append(STAGES == ['Perceive','Understand','Plan','Execute','Track','Recover'])
```

```python
world = GreenhouseWorld.demo_row(n=5, seed=2)
dets = model_perception(world, rng=np.random.default_rng(2))
targets, current = understand(dets, world)
print('detections:', len(dets), '| ranked targets:', len(targets),
      '| current:', current['id'] if current else None)
checks.append(current is None or current['reachable'])
```

### Challenge preview — all unripe
If every fruit is unripe, Understand correctly returns `current_target=None`. That is a *correct* result owned by Understand, not a failure.

```python
for f in world.fruit:
    f.ripe = False
dets = model_perception(world, rng=np.random.default_rng(2))
_, current_unripe = understand(dets, world)
print('all-unripe -> current_target:', current_unripe, '(correct: nothing to pick)')
checks.append(current_unripe is None)
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
