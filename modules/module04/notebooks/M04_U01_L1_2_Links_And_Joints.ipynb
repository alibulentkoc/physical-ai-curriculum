```python
def describe_arm(joints):
    # joints: list of (kind, link_length)
    dof=len(joints)
    config=['theta%d'%(i+1) if k=='revolute' else 'd%d'%(i+1) for i,(k,_) in enumerate(joints)]
    return dof, config
print(describe_arm([('revolute',0.4),('revolute',0.3)]))        # planar elbow arm
print(describe_arm([('revolute',0.4),('revolute',0.3),('prismatic',0.2)]))  # SCARA-like
```
