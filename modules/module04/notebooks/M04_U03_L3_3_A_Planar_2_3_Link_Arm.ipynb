## Unit 3 — Chaining Transforms (Two and Three Links)
### Lesson 3.3 — A Planar 2-/3-Link Arm

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def fk_planar(joints):
    # joints: list of (theta, L); accumulate angles, sum reaches
    phi=0.0; x=0.0; y=0.0
    for theta,L in joints:
        phi=phi+theta
        x=x+L*np.cos(phi); y=y+L*np.sin(phi)
    return np.array([x,y]), phi
js=[(np.radians(30),0.4),(np.radians(60),0.3),(np.radians(-30),0.2)]
p,ori=fk_planar(js)
print('gripper',np.round(p,3),'orientation(deg)',round(np.degrees(ori)))
```

## Coding Exercise (§8)
Verify the 3-link worked example and agreement with the matrix product.

```python
# YOUR CODE HERE
import numpy as np
assert np.allclose(p,[0.446,0.673],atol=1e-3) and np.isclose(np.degrees(ori),60)
def pose_link(theta,L):
    c,s=np.cos(theta),np.sin(theta); return np.array([[c,-s,L*c],[s,c,L*s],[0,0,1.0]])
T=np.eye(3)
for theta,L in js: T=T@pose_link(theta,L)
assert np.allclose(T[:2,2],p,atol=1e-9)
print('sum-of-reaches == matrix product (3 links).')
```
