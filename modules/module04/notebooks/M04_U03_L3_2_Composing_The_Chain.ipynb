### Lesson 3.2 — Composing the Chain

```python
import numpy as np
def pose_link(theta,L):
    c,s=np.cos(theta),np.sin(theta)
    return np.array([[c,-s,L*c],[s,c,L*s],[0,0,1.0]])
def fk_chain(joints):
    T=np.eye(3)
    for theta,L in joints:
        T=T@pose_link(theta,L)
    return T
L1,L2=0.4,0.3
T=fk_chain([(np.radians(30),L1),(np.radians(60),L2)])
print('translation',np.round(T[:2,2],3),'rotation(deg)',round(np.degrees(np.arctan2(T[1,0],T[0,0]))))
```
