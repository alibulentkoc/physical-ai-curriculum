### Lesson 1.3 — Configuration vs. Pose

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def fk_one_joint(theta, L):
    return {'pos':np.array([L*np.cos(theta),L*np.sin(theta)]),'angle':theta}
L=0.5
a=fk_one_joint(np.radians(30),L); b=fk_one_joint(np.radians(150),L)
print('config 30deg -> pose', np.round(a['pos'],3), 'angle', round(np.degrees(a['angle'])))
print('config 150deg -> pose', np.round(b['pos'],3), 'angle', round(np.degrees(b['angle'])))
```

## Coding Exercise (§8)
Show two configurations give two distinct poses (same height, different point).

```python
# YOUR CODE HERE
import numpy as np
assert np.isclose(a['pos'][1], b['pos'][1])         # same y height
assert not np.isclose(a['pos'][0], b['pos'][0])     # different x
print('joint space -> task space; distinct configs, distinct poses.')
```
