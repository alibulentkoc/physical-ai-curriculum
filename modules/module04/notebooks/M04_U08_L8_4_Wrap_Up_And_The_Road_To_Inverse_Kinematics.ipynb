```python
import numpy as np
from functools import reduce
def dh(theta,d,a,al):
    ct,st,ca,sa=np.cos(theta),np.sin(theta),np.cos(al),np.sin(al)
    return np.array([[ct,-st*ca,st*sa,a*ct],[st,ct*ca,-ct*sa,a*st],[0,sa,ca,d],[0,0,0,1.0]])
def dh_fk(table,cfg):
    return reduce(lambda A,B:A@B,[dh(q,d,a,al) for (d,a,al,_),q in zip(table,cfg)],np.eye(4))
def se3(R,t):
    T=np.eye(4);T[:3,:3]=R;T[:3,3]=t;return T
def apply(M,P): return (M@np.r_[np.asarray(P,float),1.0])[:3]
arm3=[(0.1,0,np.radians(90),'r'),(0,0.4,0,'r'),(0,0.3,0,'r')]
# full capstone on a fresh configuration and fruit
cfg=[np.radians(20),np.radians(40),np.radians(50)]
T=dh_fk(arm3,cfg)
Tbc=se3(np.eye(3),[0.1,0,0.5]); P_base=apply(Tbc,[0.04,0.02,0.42])
move=P_base-T[:3,3]
print('gripper',np.round(T[:3,3],3),' fruit',np.round(P_base,3),' move',np.round(move,3))
```
