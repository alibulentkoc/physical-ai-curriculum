```python
import numpy as np
def tip_position(theta, L):
    return np.array([L*np.cos(theta), L*np.sin(theta)])
L=0.5
for deg in [0,90,180]:
    print(f'theta={deg:>3}deg -> tip', np.round(tip_position(np.radians(deg),L),3))
```

## Coding Exercise (§8)
Confirm the worked-example tip positions for the one-joint arm.

```python
# YOUR CODE HERE
import numpy as np
assert np.allclose(tip_position(0,L),[0.5,0])
assert np.allclose(tip_position(np.radians(90),L),[0,0.5])
print('forward kinematics (1 joint): angle -> position works.')
```
