### Lesson 2.3 — Where Is the Tip?

```python
import numpy as np
def pose_one_joint(theta,L):
    c,s=np.cos(theta),np.sin(theta); return np.array([[c,-s,L*c],[s,c,L*s],[0,0,1.0]])
def position_of(T): return T[:2,2]
def orientation_of(T): return np.degrees(np.arctan2(T[1,0],T[0,0]))
L=0.5
for deg in [90,210]:
    T=pose_one_joint(np.radians(deg),L)
    print(f'theta={deg} pos',np.round(position_of(T),3),'orient',round(orientation_of(T)))
```

## Coding Exercise (§8)
Extract position/orientation and cross-check with trig.

```python
# YOUR CODE HERE
import numpy as np
T=pose_one_joint(np.radians(210),L)
assert np.allclose(position_of(T),[0.5*np.cos(np.radians(210)),0.5*np.sin(np.radians(210))],atol=1e-3)
assert abs(((orientation_of(T)-210+180)%360)-180)<1e-6
print('position = translation column; orientation = rotation block; matches trig.')
```
