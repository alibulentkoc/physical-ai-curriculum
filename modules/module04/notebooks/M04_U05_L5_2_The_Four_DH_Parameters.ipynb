```python
# Fill the joint variable into each row, given a configuration.
def dh_row(theta,d,a,alpha,kind='revolute'):
    return {'theta':theta,'d':d,'a':a,'alpha':alpha,'kind':kind}
def fill(table,config):
    out=[]
    for r,q in zip(table,config):
        r2=dict(r)
        if r2['kind']=='revolute': r2['theta']=q
        else: r2['d']=q
        out.append(r2)
    return out
import numpy as np
planar=[dh_row(0,0,0.4,0),dh_row(0,0,0.3,0)]
filled=fill(planar,[np.radians(30),np.radians(60)])
print('filled thetas (deg):',[round(np.degrees(r['theta'])) for r in filled])
```

Verify the variable goes into theta (revolute) or d (prismatic).

```python
# YOUR CODE HERE
import numpy as np
assert np.isclose(filled[0]['theta'],np.radians(30))
pr=[dh_row(0,0,0,0,'prismatic')]
fp=fill(pr,[0.25])
assert fp[0]['d']==0.25 and fp[0]['theta']==0
print('revolute fills theta; prismatic fills d.')
```
