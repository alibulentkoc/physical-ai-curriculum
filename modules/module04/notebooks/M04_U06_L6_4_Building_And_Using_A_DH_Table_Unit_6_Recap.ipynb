```python
import numpy as np
from functools import reduce
def dh_transform(theta,d,a,alpha):
    ct,st,ca,sa=np.cos(theta),np.sin(theta),np.cos(alpha),np.sin(alpha)
    return np.array([[ct,-st*ca,st*sa,a*ct],[st,ct*ca,-ct*sa,a*st],[0,sa,ca,d],[0,0,0,1.0]])
def dh_fk(table,config):
    fac=[]
    for (d,a,alpha,kind),q in zip(table,config):
        theta=q if kind=='revolute' else 0.0
        dd=q if kind=='prismatic' else d
        fac.append(dh_transform(theta,dd,a,alpha))
    return reduce(lambda A,B:A@B,fac,np.eye(4))
arm3=[(0.1,0,np.radians(90),'revolute'),(0,0.4,0,'revolute'),(0,0.3,0,'revolute')]
T=dh_fk(arm3,[0,np.radians(30),np.radians(60)])
print('position',np.round(T[:3,3],3))
print('orientation R=\n',np.round(T[:3,:3],3))
```

## Coding Exercise (§8)
Compute T_0^3, extract position/orientation, confirm the in-plane reach matches the planar formula.

```python
# YOUR CODE HERE
import numpy as np
p=T[:3,3]
reach=np.hypot(np.hypot(p[0],p[1]), p[2]-0.1)
planar=0.4*np.cos(np.radians(30))+0.3*np.cos(np.radians(90)),0.4*np.sin(np.radians(30))+0.3*np.sin(np.radians(90))
assert np.isclose(reach, np.hypot(*planar), atol=1e-6)
assert T[:3,:3].shape==(3,3)
print('table -> dh_fk -> pose; in-plane reach matches planar formula.')
```
