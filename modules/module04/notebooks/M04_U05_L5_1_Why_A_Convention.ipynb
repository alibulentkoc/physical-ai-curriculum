```python
# A joint's DH parameters as a record; the planar 2-link arm as a table.
def dh_row(theta,d,a,alpha,kind='revolute'):
    return {'theta':theta,'d':d,'a':a,'alpha':alpha,'kind':kind}
planar=[dh_row(0.0,0,0.4,0),dh_row(0.0,0,0.3,0)]   # theta is the variable per row
for i,r in enumerate(planar,1):
    print(f"row {i}: theta={r['theta']} d={r['d']} a={r['a']} alpha={r['alpha']} ({r['kind']})")
```

## Coding Exercise (§8)
Confirm four parameters per joint and that the planar arm has a=L, others zero (theta variable).

```python
# YOUR CODE HERE
assert all(len(r)==5 for r in planar)
assert planar[0]['a']==0.4 and planar[1]['a']==0.3
assert all(r['d']==0 and r['alpha']==0 for r in planar)
print('four DH parameters per joint; planar arm encoded as a table.')
```
