```python
import numpy as np
# 3-DOF capstone arm DH table (standard); theta entries are variables.
def row(d,a,alpha,kind='revolute'): return {'d':d,'a':a,'alpha':alpha,'kind':kind}
table=[row(0.1,0,np.radians(90)),row(0,0.4,0),row(0,0.3,0)]
def fill(table,config):
    out=[]
    for r,q in zip(table,config):
        rr=dict(r); rr['theta']=q if rr['kind']=='revolute' else 0.0
        if rr['kind']=='prismatic': rr['d']=q
        out.append(rr)
    return out
filled=fill(table,[np.radians(0),np.radians(30),np.radians(60)])
print('a column:',[r['a'] for r in filled],' alpha1(deg):',round(np.degrees(filled[0]['alpha'])))
```
