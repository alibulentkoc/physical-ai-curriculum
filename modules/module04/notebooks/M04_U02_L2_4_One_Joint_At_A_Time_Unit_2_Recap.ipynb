```python
import numpy as np
def pose_one_joint(theta,L):
    c,s=np.cos(theta),np.sin(theta); return np.array([[c,-s,L*c],[s,c,L*s],[0,0,1.0]])
L=0.5
for deg in [0,120]:
    T=pose_one_joint(np.radians(deg),L)
    print(f'theta={deg} pos',np.round(T[:2,2],3),'orient',round(np.degrees(np.arctan2(T[1,0],T[0,0]))))
```
