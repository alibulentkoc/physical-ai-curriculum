```python
def dh_row(theta,d,a,alpha,kind='revolute'):
    return {'theta':theta,'d':d,'a':a,'alpha':alpha,'kind':kind}
def show(table):
    print('i  theta  d     a     alpha  kind')
    for i,r in enumerate(table,1):
        print(f"{i}  {r['theta']:<5} {r['d']:<5} {r['a']:<5} {r['alpha']:<6} {r['kind']}")
arm3=[dh_row('θ1',0.1,0,90),dh_row('θ2',0,0.4,0),dh_row('θ3',0,0.3,0)]
show(arm3)
```

## Coding Exercise (§8)
Encode the 3-DOF arm table and confirm its constant columns.

```python
# YOUR CODE HERE
assert arm3[0]['alpha']==90 and arm3[0]['d']==0.1
assert arm3[1]['a']==0.4 and arm3[2]['a']==0.3
assert all(r['kind']=='revolute' for r in arm3)
print('3-DOF DH table encoded; constants verified.')
```
