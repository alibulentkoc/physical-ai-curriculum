### Lesson 4.5 — Applying SE(3); Inverses in 3D

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def se3_inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
T=se3(Rz(90),[1,0,0]); Ti=se3_inv(T)
print('apply T to (1,0,0):', np.round((T@np.array([1,0,0,1.]))[:3],3))
print('T_inv @ T = I:', np.allclose(Ti@T,np.eye(4)))
```

## Coding Exercise (§8)
Apply T then T_inv to several 3D points; recover the originals; confirm distances preserved.

```python
# YOUR CODE HERE
import numpy as np
pts=np.array([[1,0,0,1.],[0,1,0,1.],[0.5,0.5,1,1.]]).T
back=Ti@(T@pts)
print('recovered originals:', np.allclose(back,pts))
import itertools
def pw(P): return [round(np.linalg.norm(P[:3,i]-P[:3,j]),4) for i,j in itertools.combinations(range(P.shape[1]),2)]
print('distances preserved:', pw(pts)==pw(T@pts))
```
