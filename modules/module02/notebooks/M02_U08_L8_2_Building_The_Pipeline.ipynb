```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])

def tomato_world_pose(T_world_arm, T_arm_cam, T_cam_tomato):
    return T_world_arm @ T_arm_cam @ T_cam_tomato

I=np.eye(3)
T_cam_tomato = se3(I,[0,0,0.3])
T_arm_cam    = se3(I,[0,0,0.1])
T_world_arm  = se3(I,[1.0,0.5,0])
T = tomato_world_pose(T_world_arm,T_arm_cam,T_cam_tomato)
print('world position   :', np.round(T[:3,3],3))
print('world orientation:\n', np.round(T[:3,:3],3))
```

## Coding Exercise (§8)
Apply the chain to the tomato-frame origin to get the world point; add a detection rotation and watch the orientation block change.

```python
# YOUR CODE HERE
import numpy as np
origin=np.array([0,0,0,1.0])
print('tomato world point:', np.round((T@origin)[:3],3))
# add a 30-degree detection rotation
T_cam_tomato_rot = se3(Rz(30),[0,0,0.3])
T2 = tomato_world_pose(T_world_arm,T_arm_cam,T_cam_tomato_rot)
print('world orientation now carries the rotation:\n', np.round(T2[:3,:3],3))
```
