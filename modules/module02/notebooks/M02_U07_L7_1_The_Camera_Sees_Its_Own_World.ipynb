### Lesson 7.1 — The Camera Sees Its Own World

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
# camera's pose in the world (no rotation, 1.0 forward, 0.5 up)
T_world_cam=se3(np.eye(3),[1.0,0.0,0.5])
p_cam=np.array([0.3,0,0,1.0])   # 0.3 m ahead in camera frame
p_world=T_world_cam@p_cam
print('detection in camera frame:', p_cam[:3])
print('detection in world frame :', np.round(p_world[:3],3))
```

## Coding Exercise (§8)
Change only the camera pose; show the world result changes but the camera-frame value does not.

```python
# YOUR CODE HERE
import numpy as np
T2=se3(np.eye(3),[2.0,1.0,0.5])
print('camera-frame value (unchanged):', p_cam[:3])
print('world with pose A:', np.round((T_world_cam@p_cam)[:3],3))
print('world with pose B:', np.round((T2@p_cam)[:3],3))
```
