### Lesson 6.1 — What Is a Pose?

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
pose=se3(Rz(90),[0.4,0.3,0.9])
print('pose (one matrix):\n', np.round(pose,3))
print('position:', np.round(pose[:3,3],3))
print('orientation block:\n', np.round(pose[:3,:3],3))
```

## Coding Exercise (§8)
Show two poses with the same position but different orientation are different matrices.

```python
# YOUR CODE HERE
import numpy as np
a=se3(Rz(0),[0.4,0.3,0.9]); b=se3(Rz(90),[0.4,0.3,0.9])
print('same position:', np.allclose(a[:3,3],b[:3,3]))
print('same pose:', np.allclose(a,b))
```
