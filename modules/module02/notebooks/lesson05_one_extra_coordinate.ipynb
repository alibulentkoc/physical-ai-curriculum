# Lesson 2.1 Notebook — One Extra Coordinate
Convert points to/from homogeneous form. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def to_h(p): return np.array([p[0], p[1], 1.0])     # (x,y) -> (x,y,1)
def from_h(h): return np.array([h[0], h[1]])         # drop last entry
p=np.array([2.0,3.0])
print('homogeneous:', to_h(p))
I3=np.eye(3)
print('identity 3x3 changes nothing:', from_h(I3 @ to_h(p)))
```

## Coding Exercise (§8)
Round-trip several points through homogeneous form and back; confirm unchanged.

```python
# YOUR CODE HERE
import numpy as np
pts=[np.array([0.,0.]),np.array([2.,3.]),np.array([-1.,4.])]
ok=all(np.allclose(from_h(to_h(q)),q) for q in pts)
print('round-trip preserved:', ok)
```
