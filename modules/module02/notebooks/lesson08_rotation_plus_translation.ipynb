# Lesson 2.4 Notebook — Rotation + Translation in One Matrix
Block form [[R, t],[0, 1]]. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def M(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
m=M(90,1,0)
print('M(90,1,0) =\n', np.round(m,3))
print('apply to (1,0,1):', np.round(m@np.array([1,0,1.]),3))   # rotate then offset -> (1,1,1)
```

## Coding Exercise (§8)
Confirm the one-matrix result equals rotate-then-translate done in two steps.

```python
# YOUR CODE HERE
import numpy as np
def R(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.]])
def T(tx,ty): return np.array([[1,0,tx],[0,1,ty],[0,0,1.]])
p=np.array([1,0,1.])
one_matrix = M(90,1,0)@p
two_step   = (T(1,0)@R(90))@p
print('equal:', np.allclose(one_matrix, two_step))
```
