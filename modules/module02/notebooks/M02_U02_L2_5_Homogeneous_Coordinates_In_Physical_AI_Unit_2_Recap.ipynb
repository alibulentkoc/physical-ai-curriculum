```python
import numpy as np
def M(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
point=np.array([2.,0.,1.]); direction=np.array([2.,0.,0.])
m=M(90,1,1)
print('point     ->', np.round(m@point,3))
print('direction ->', np.round(m@direction,3), '(rotated, not translated)')
```

## Coding Exercise (§8)
Chain two combined transforms by multiplication and apply to a point.

```python
# YOUR CODE HERE
import numpy as np
chain = M(90,1,0) @ M(0,2,0)    # move +2 x, then rotate 90 + move +1 x
print('chained matrix @ (0,0,1):', np.round(chain@np.array([0,0,1.]),3))
```
