One tomato, three frames, by hand (no transforms yet). Run **Kernel → Restart & Run All**.

```python
import numpy as np
# A tomato described in three frames (axes aligned here, for intuition only)
camera = np.array([0.30, 0.00])   # ahead of the lens
# arm is 0.05 behind, 0.25 below the camera -> add that offset
arm    = camera + np.array([-0.05, -0.25])
# base sits at world (1.65, 1.35)
world  = arm + np.array([1.65, 1.35])
print('camera frame:', camera)
print('arm frame   :', arm)
print('world frame :', world)
print('Same tomato. Three descriptions. The robot restates it constantly.')
```
