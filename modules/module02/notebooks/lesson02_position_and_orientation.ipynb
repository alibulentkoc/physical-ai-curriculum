# Lesson 1.2 Notebook — Position and Orientation Travel Together
A pose = (x, y, theta). Run **Kernel → Restart & Run All**.

```python
import numpy as np
def pose(x,y,theta_deg): return {'x':x,'y':y,'theta':theta_deg}
def poses_equal(a,b): return a['x']==b['x'] and a['y']==b['y'] and a['theta']==b['theta']
grasp = pose(0.35,0.15,120)
print('grasp pose:', grasp)
# same position, different orientation -> different pose
other = pose(0.35,0.15,0)
print('same position, different orientation equal?', poses_equal(grasp,other))
```

## Coding Exercise (§8)
Show two poses are equal only if BOTH position and orientation match.

```python
# YOUR CODE HERE
a=pose(1,2,90); b=pose(1,2,90); c=pose(1,2,45)
print('a == b:', poses_equal(a,b), '  a == c:', poses_equal(a,c))
```
