```python
import numpy as np
# A 2D (floor) pose: position + ONE heading angle
pose2d={'x':0.4,'y':0.3,'theta_deg':30}
# A 3D pose: position (x,y,z) + THREE orientation numbers
pose3d={'x':0.4,'y':0.3,'z':0.9,'roll':0,'pitch':20,'yaw':30}
print('2D pose numbers:', len(pose2d), '->', pose2d)
print('3D pose numbers:', len(pose3d), '->', pose3d)
print('3D adds height z and two more orientation axes.')
```
