# Lesson 2.2 Notebook — Points vs Directions
w=1 point translates; w=0 direction doesn't. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def T(tx,ty): return np.array([[1,0,tx],[0,1,ty],[0,0,1.0]])
point     = np.array([2.0,3.0,1.0])   # w=1
direction = np.array([2.0,3.0,0.0])   # w=0
M=T(5,0)
print('point     after T(5,0):', M@point,     '(moved)')
print('direction after T(5,0):', M@direction, '(unchanged)')
```
