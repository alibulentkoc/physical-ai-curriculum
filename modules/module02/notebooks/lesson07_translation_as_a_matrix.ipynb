# Lesson 2.3 Notebook — Translation as a Matrix
T(tx,ty) with offset in the last column. Run **Kernel → Restart & Run All**.

```python
import numpy as np
def T(tx,ty): return np.array([[1,0,tx],[0,1,ty],[0,0,1.0]])
print('T(5,-1) =\n', T(5,-1))
print('apply to (2,3,1):', T(5,-1)@np.array([2,3,1.]))   # (7,2,1)
# two translations compose by adding offsets
print('T(5,-1)@T(2,2) == T(7,1):', np.allclose(T(5,-1)@T(2,2), T(7,1)))
```
