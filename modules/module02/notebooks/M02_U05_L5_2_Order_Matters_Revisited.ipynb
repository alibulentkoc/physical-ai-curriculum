```python
import numpy as np
def se2(d,tx,ty):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
R=se2(90,0,0); T=se2(0,3,0)
o=np.array([0,0,1.0])
print('T*R applied to origin:', np.round((T@R)@o,3))  # rotate then translate -> (3,0)
print('R*T applied to origin:', np.round((R@T)@o,3))  # translate then rotate -> (0,3)
print('orders differ:', not np.allclose((T@R)@o,(R@T)@o))
```

## Coding Exercise (§8)
Show two pure translations DO commute, but a rotation and a translation do not.

```python
# YOUR CODE HERE
import numpy as np
Ta=se2(0,1,2); Tb=se2(0,3,-1)
print('translations commute:', np.allclose(Ta@Tb, Tb@Ta))
print('rotation+translation commute:', np.allclose(R@T, T@R))
```
