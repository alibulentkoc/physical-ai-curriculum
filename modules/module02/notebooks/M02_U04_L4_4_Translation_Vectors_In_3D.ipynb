## Unit 4 — SE(3) Transformations
### Lesson 4.4 — Translation Vectors in 3D

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
T=se3(np.eye(3),[0.4,0.3,0.9])  # translation only
point=np.array([0,0,0,1.0]); direction=np.array([0,0,1,0.0])
print('point     ->', np.round(T@point,3), '(moved to t)')
print('direction ->', np.round(T@direction,3), '(unchanged: w=0)')
```

## Coding Exercise (§8)
Add a rotation and observe both rotate; only the point also translates.

```python
# YOUR CODE HERE
import numpy as np
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
T2=se3(Rz(90),[0.4,0.3,0.9])
print('point     ->', np.round(T2@np.array([1,0,0,1.]),3))
print('direction ->', np.round(T2@np.array([1,0,0,0.]),3), '(rotated, not translated)')
```
