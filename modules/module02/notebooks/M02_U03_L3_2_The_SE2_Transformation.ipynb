### Lesson 3.2 — The SE(2) Transformation

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se2(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def read_se2(T):
    theta=np.degrees(np.arctan2(T[1,0],T[0,0])); return round(theta,2),round(T[0,2],3),round(T[1,2],3)
T=se2(90,2,1)
print('T =\n',np.round(T,3))
print('read back (theta,tx,ty):', read_se2(T))
```

Round-trip several poses: build from (theta,tx,ty), read them back.
