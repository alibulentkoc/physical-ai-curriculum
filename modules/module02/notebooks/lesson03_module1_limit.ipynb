Every 2x2 matrix fixes the origin. Run **Kernel → Restart & Run All**.

```python
import numpy as np
origin=np.array([0.0,0.0])
mats={'rotate90':np.array([[0,-1],[1,0]]),'scale':np.array([[2,0],[0,0.5]]),'reflect':np.array([[1,0],[0,-1]])}
for name,M in mats.items():
    print(f'{name:9s} @ origin =', M@origin)
print('Every 2x2 matrix maps (0,0) -> (0,0): the origin is fixed.')
```

## Coding Exercise (§8)
Translation must send (0,0) somewhere new — show no 2x2 matrix can do that; you must ADD a vector.

```python
# YOUR CODE HERE
import numpy as np
t=np.array([1.0,0.0])     # desired: move right by 1
print('translate via addition:', origin + t)   # works, but is NOT a 2x2 multiply
print('any 2x2 @ origin stays at origin -> cannot translate with a 2x2 matrix')
```
