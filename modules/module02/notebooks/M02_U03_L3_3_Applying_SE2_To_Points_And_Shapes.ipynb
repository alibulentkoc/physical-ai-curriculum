### Lesson 3.3 — Applying SE(2) to Points and Shapes

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se2(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def apply(T,pts):
    h=np.vstack([pts,np.ones(pts.shape[1])]); return (T@h)[:2]
T=se2(90,2,1)
print('apply T to (1,0):', np.round(apply(T,np.array([[1],[0.0]]))[:,0],3))
print('apply T to (0,0):', np.round(apply(T,np.array([[0],[0.0]]))[:,0],3))
```

## Coding Exercise (§8)
Transform a unit square and confirm pairwise distances are preserved.

```python
# YOUR CODE HERE
import numpy as np, itertools
sq=np.array([[0,1,1,0],[0,0,1,1.0]])
out=apply(T,sq)
def pw(p): return [round(np.linalg.norm(p[:,i]-p[:,j]),4) for i,j in itertools.combinations(range(p.shape[1]),2)]
print('distances preserved:', pw(sq)==pw(out))
```
