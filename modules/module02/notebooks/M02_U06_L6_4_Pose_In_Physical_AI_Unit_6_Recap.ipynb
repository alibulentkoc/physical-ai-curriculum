```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
I=np.eye(3)
T_w_b=se3(I,[1,0,0]); T_b_g=se3(I,[0,0,0.5])
T_w_g=T_w_b@T_b_g
print('gripper in world position:', np.round(T_w_g[:3,3],3))
# update base pose with a motion, recompute gripper-in-world
T_w_b2=T_w_b@se3(I,[0.5,0,0])
print('after base moves, gripper in world:', np.round((T_w_b2@T_b_g)[:3,3],3))
```
