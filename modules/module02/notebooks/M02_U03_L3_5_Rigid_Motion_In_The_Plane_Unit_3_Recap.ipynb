```python
import numpy as np, itertools
def se2(d,tx,ty):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def se2_inv(T):
    R=T[:2,:2]; t=T[:2,2]; Ti=np.eye(3); Ti[:2,:2]=R.T; Ti[:2,2]=-R.T@t; return Ti
def apply(T,pts): h=np.vstack([pts,np.ones(pts.shape[1])]); return (T@h)[:2]
def pw(p): return [round(np.linalg.norm(p[:,i]-p[:,j]),4) for i,j in itertools.combinations(range(p.shape[1]),2)]
sq=np.array([[0,1,1,0],[0,0,1,1.0]]); T=se2(35,1.2,0.6)
moved=apply(T,sq); back=apply(se2_inv(T),moved)
print('rigid (distances preserved):', pw(sq)==pw(moved))
print('inverse returns to start   :', np.allclose(back,sq))
```
