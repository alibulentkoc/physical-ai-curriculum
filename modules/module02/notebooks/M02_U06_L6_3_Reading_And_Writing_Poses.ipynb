```python
import numpy as np
def se2(d,tx,ty):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def read(T): return round(np.degrees(np.arctan2(T[1,0],T[0,0])),2), round(T[0,2],3), round(T[1,2],3)
old=se2(0,2,1)
delta=se2(90,1,0)   # in robot frame: forward 1 then turn 90
new=old@delta       # right-multiply: motion in robot's own frame
print('old (theta,x,y):', read(old))
print('new (theta,x,y):', read(new))
```

## Coding Exercise (§8)
Apply a world-frame motion by left-multiplication and compare.

```python
# YOUR CODE HERE
import numpy as np
world_move=se2(0,0,1)   # world-frame: shift +y by 1
print('robot-frame update:', read(old@world_move))
print('world-frame update:', read(world_move@old))
```
