```python
import numpy as np
def se2(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def se2_inv(T):
    R=T[:2,:2]; t=T[:2,2]
    Ti=np.eye(3); Ti[:2,:2]=R.T; Ti[:2,2]=-R.T@t; return Ti
T=se2(90,2,1); Ti=se2_inv(T)
print('T_inv @ T = identity:', np.allclose(Ti@T,np.eye(3)))
p=np.array([2,1,1.0])
print('T_inv applied to (2,1):', np.round((Ti@p)[:2],3), '(back to start)')
```
