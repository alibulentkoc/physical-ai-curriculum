```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
I=np.eye(3)
T_cam_tomato=se3(I,[0,0,0.3]); T_arm_cam=se3(I,[0,0,0.1]); T_world_arm=se3(I,[1.0,0.5,0])
T_world_cam=T_world_arm@T_arm_cam
T_world_tomato=T_world_cam@T_cam_tomato
# Round-trip: inverse chain recovers the detection
recovered = inv(T_world_cam) @ T_world_tomato
print('round-trip recovers detection:', np.allclose(recovered, T_cam_tomato))
# Distance check
d_cam = np.linalg.norm(T_cam_tomato[:3,3])
d_world = np.linalg.norm(T_world_tomato[:3,3]-T_world_cam[:3,3])
print('distance preserved:', np.isclose(d_cam,d_world), '(', round(d_cam,3),'m )')
```

## Coding Exercise (§8)
Assert the round-trip and distance checks, validate every matrix is a proper pose, and print the four frame origins.

```python
# YOUR CODE HERE
import numpy as np
def is_pose(T):
    return np.allclose(T[3],[0,0,0,1]) and np.allclose(T[:3,:3]@T[:3,:3].T,np.eye(3)) and np.isclose(np.linalg.det(T[:3,:3]),1)
for name,T in [('world',np.eye(4)),('arm',T_world_arm),('cam',T_world_cam),('tomato',T_world_tomato)]:
    print(f'{name:7s} origin = {np.round(T[:3,3],3)}  valid pose: {is_pose(T)}')
```
