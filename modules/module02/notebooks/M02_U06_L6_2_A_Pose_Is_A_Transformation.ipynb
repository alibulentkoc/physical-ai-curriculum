### Lesson 6.2 — A Pose Is a Transformation

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
I=np.eye(3)
T_w_b=se3(I,[1,0,0])     # base pose in world
T_b_g=se3(I,[0,0,0.5])   # gripper pose in base
T_w_g=T_w_b@T_b_g        # gripper pose in world
print('gripper in world:\n', np.round(T_w_g,3))
p_g=np.array([0,0,0,1.0])
print('gripper origin in world:', np.round((T_w_g@p_g)[:3],3))
```

## Coding Exercise (§8)
Use the inverse to bring a world point into the gripper frame.

```python
# YOUR CODE HERE
import numpy as np
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
world_pt=np.array([1,0,0.5,1.0])
print('world point in gripper frame:', np.round((inv(T_w_g)@world_pt)[:3],3))
```
