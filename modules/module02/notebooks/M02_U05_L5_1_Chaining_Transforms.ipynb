### Lesson 5.1 — Chaining Transforms

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se2(d,tx,ty):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
T1=se2(90,0,0)   # rotate 90
T2=se2(0,2,0)    # translate (2,0)
p=np.array([1,0,1.0])
step=T2@(T1@p)
chain=(T2@T1)@p
print('step-by-step:', np.round(step,3))
print('one matrix   :', np.round(chain,3))
print('equal:', np.allclose(step,chain))
```
