# Lesson 2.1 — The Twin's State: Mirroring the World-State

The twin holds a **copyable** frame of the reported world-state. Capturing it and then mutating the real world must NOT change the captured frame.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
real.q = np.array([0.5, 0.7]); real.fruit[0].picked = True
frame = snapshot(real)
for key in ('q','tool_xy','fruit','health','stage'):
    checks.append(key in frame)
print('frame keys:', sorted(frame.keys()))
# the frame is a COPY: mutate the world afterward, frame is unchanged
q_before = frame['q'].copy(); picked_before = frame['fruit']['F0']['picked']
real.q = np.array([1.2, -0.3]); real.fruit[0].picked = False
checks.append(np.allclose(frame['q'], q_before))               # q copy held
checks.append(frame['fruit']['F0']['picked'] == picked_before) # fruit copy held
print('frame is a self-contained copy (unchanged after world mutated):', checks[-2] and checks[-1])
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
