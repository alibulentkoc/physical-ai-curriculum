# Lesson 5.1 — Monitoring as Reality ↔ Twin Comparison

A monitor answers 'what is happening now?' by comparing reality against the twin. Divergence is the signal.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, model_perception, understand,
                                          run_pipeline)
from modules.module10.twin import (DigitalTwin, GroundTruth, outcome_gap,
                                   monitor, predict, compare_futures)
def victim_and_xy(seed=1):
    w = GreenhouseWorld.demo_row(n=6, seed=seed)
    dets = model_perception(w, rng=np.random.default_rng(7))
    v = understand(dets, w)[1]['id']
    vxy = next(d['xy'] for d in dets if d['id']==v)
    return v, vxy
checks = []
ground = GroundTruth(GreenhouseWorld.demo_row(n=6, seed=1))
twin = DigitalTwin(ground.world); twin.sync(ground.report())
m0 = monitor(twin, ground.report())
print('after sync -> alert=%s (%s)' % (m0['alert'], m0['summary']))
checks.append(m0['alert'] is False and m0['synced'])      # in sync -> quiet
# reality moves on; twin not re-synced -> divergence signal
ground.world.q = ground.world.q + np.array([0.10, -0.05])
m1 = monitor(twin, ground.report())
print('reality moved -> alert=%s q_gap=%.3f' % (m1['alert'], m1['q_gap']))
checks.append(m1['alert'] is True and m1['q_gap'] > 0)    # divergence = signal
twin.sync(ground.report())
checks.append(monitor(twin, ground.report())['alert'] is False)   # re-sync clears it
print('re-sync clears the signal:', not monitor(twin, ground.report())['alert'])
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
