# Lesson 3.2 — The Twin as a Sandbox

Inject a what-if fault and simulate it safely on a copy. The experiment is non-destructive: a fresh simulation still predicts the clean harvest.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, model_perception, understand
from modules.module10.twin import DigitalTwin, GroundTruth, outcome_gap, snapshot
def victim_and_xy(seed=1):
    w = GreenhouseWorld.demo_row(n=6, seed=seed)
    dets = model_perception(w, rng=np.random.default_rng(7))
    v = understand(dets, w)[1]['id']
    vxy = next(d['xy'] for d in dets if d['id']==v)
    return v, vxy
checks = []
v, vxy = victim_and_xy()
twin = DigitalTwin(GreenhouseWorld.demo_row(n=6, seed=1))
blocked = twin.simulate(inject={v: dict(obstacle=(vxy, 0.25))})   # rehearse a blocked fruit
print('with obstacle on %s -> skipped=%s harvested=%d' % (v, blocked['skipped'], len(blocked['harvested'])))
checks.append(v in blocked['skipped'])                            # that fruit skipped
checks.append(len(blocked['harvested']) >= 1)                     # the rest harvested
clean = twin.simulate()                                          # non-destructive: fresh run is clean
print('fresh simulation (no inject) -> %s harvested, %s skipped' % (len(clean['harvested']), clean['skipped']))
checks.append(v in clean['harvested'])                            # the what-if didn't persist
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
