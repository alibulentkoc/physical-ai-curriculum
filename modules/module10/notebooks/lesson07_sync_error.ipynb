# Lesson 2.3 — When the Mirror Drifts: Sync Error

Reality may carry an **unmodeled effect** it does not report. A perfect sync to the *reported* state still leaves a residual gap to the *true* state.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
ground = GroundTruth(GreenhouseWorld.demo_row(n=6, seed=1), q_offset=np.array([0.05,-0.03]))
twin = DigitalTwin(ground.world)
twin.sync(ground.report())                       # sync to the REPORTED state
sync_err = twin.divergence(ground.report())      # twin vs report  -> ~0
residual = twin.divergence(ground.true_report()) # twin vs truth   -> > 0
print('sync error (vs report): q_gap=%.4f synced=%s' % (sync_err['q_gap'], sync_err['synced']))
print('residual gap (vs truth): q_gap=%.4f' % residual['q_gap'])
checks.append(sync_err['synced'])                # flawless mirror of the report
checks.append(residual['q_gap'] > 0.01)          # but still off from reality
print('synced to the report != matching reality:', checks[0] and checks[1])
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
