# Lesson 7.1 — Pre-Validating Actions in the Twin

Rehearse a candidate action in the twin **before** committing it to reality. `prevalidate` runs the action forward and applies a stated accept/reject rule. No learning, no optimization.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, harvest_row
from modules.module10.twin import (GroundTruth, DigitalTwin, snapshot,
    prevalidate, select_action, twin_in_the_loop)
mk = lambda: GreenhouseWorld.demo_row(n=5, seed=1)
F2 = mk().fruit[2]
OBST = {'F2': {'obstacle': ((float(F2.x), float(F2.y)), 0.05)}}
def effort(rep): return sum(p['n_attempts'] for p in rep['picks'])
checks = []
twin = DigitalTwin(mk()); twin.sync(snapshot(mk()))
# A clean action: nominal plan, no obstacle -> should be ACCEPTED
clean = prevalidate(twin, action=None)
print('clean: accept =', clean['accept'], '| skipped =', clean['forecast']['skipped'])
checks.append(clean['accept'] is True)
checks.append(clean['forecast']['skipped'] == [])
```

```python
# The same pick facing an obstacle on F2 -> forecast skips F2 -> REJECTED
blocked = prevalidate(twin, action=OBST)
print('blocked: accept =', blocked['accept'], '| skipped =', blocked['forecast']['skipped'])
checks.append(blocked['accept'] is False)
checks.append('F2' in blocked['forecast']['skipped'])
```

```python
# Pre-validation is a forecast + a stated rule, run on the twin's own copy.
assert all(checks), checks
print('All checks passed.')
```
