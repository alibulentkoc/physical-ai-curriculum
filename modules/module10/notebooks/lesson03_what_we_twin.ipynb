# Lesson 1.3 — What We Twin: the Module 9 System

The twin wraps four Module 9 inputs (orchestrator, world-state, health signals, interface). We run one real pick through the interface and read the reported snapshot the twin will mirror.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
r = run_pipeline(real, real.q.copy())                 # the M9 system interface
rep = snapshot(real, health=r['monitor'], stage='executed')
for key in ('q','tool_xy','fruit','health','stage'):  # the reported world-state
    checks.append(key in rep)
print('reported state keys:', sorted(rep.keys()))
print('health signals present:', sorted(rep['health'].keys()) if rep['health'] else None)
# the twin reuses the M9 layout without redefining any layer
twin = DigitalTwin(real)
checks.append([f.fid for f in twin.world.fruit] == [f.fid for f in real.fruit])
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
