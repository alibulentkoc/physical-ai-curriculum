# Lesson 7.2 — Twin-Informed Decisions

Pre-validate several candidate actions and **choose the better-forecast one** via a simple stated score. Action-selection, not optimization or RL.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, harvest_row
from modules.module10.twin import (GroundTruth, DigitalTwin, snapshot,
    prevalidate, select_action, twin_in_the_loop)
mk = lambda: GreenhouseWorld.demo_row(n=5, seed=1)
F2 = mk().fruit[2]
OBST = {'F2': {'obstacle': ((float(F2.x), float(F2.y)), 0.05)}}
def effort(rep): return sum(p['n_attempts'] for p in rep['picks'])
checks = []
twin = DigitalTwin(mk()); twin.sync(snapshot(mk()))
candidates = {'clear_path': None, 'ignore_obstacle': OBST}
sel = select_action(twin, candidates)
print('ranking:', sel['ranking'], '| chosen:', sel['chosen'])
print('scores:', sel['scores'])
```

```python
# clear_path harvests 4 / skips 0 -> outranks ignore_obstacle (3 / 1)
checks.append(sel['chosen'] == 'clear_path')
checks.append(sel['ranking'][0] == 'clear_path')
checks.append(sel['scores']['clear_path'] > sel['scores']['ignore_obstacle'])
```

```python
# Changing the rule changes the choice predictably -> accountable, not learned.
assert all(checks), checks
print('All checks passed.')
```
