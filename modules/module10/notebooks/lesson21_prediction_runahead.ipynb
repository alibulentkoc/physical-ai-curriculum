# Lesson 6.1 — Prediction as Run-Ahead Simulation

Predict = sync to now, then run the existing system ahead. A forecast, on a copy, reality untouched, reproducible.

```python
import numpy as np
from modules.module09.integration import (GreenhouseWorld, model_perception, understand,
                                          run_pipeline)
from modules.module10.twin import (DigitalTwin, GroundTruth, outcome_gap,
                                   monitor, predict, compare_futures)
def victim_and_xy(seed=1):
    w = GreenhouseWorld.demo_row(n=6, seed=seed)
    dets = model_perception(w, rng=np.random.default_rng(7))
    v = understand(dets, w)[1]['id']
    vxy = next(d['xy'] for d in dets if d['id']==v)
    return v, vxy
checks = []
ground = GroundTruth(GreenhouseWorld.demo_row(n=6, seed=1))
twin = DigitalTwin(ground.world)
before = [f.picked for f in ground.world.fruit]
fc = predict(twin, sync_report=ground.report())     # sync-then-run-ahead
print('forecast -> harvested=%d skipped=%d complete=%s' % (len(fc['harvested']), len(fc['skipped']), fc['complete']))
for k in ('harvested','skipped','picks','complete'): checks.append(k in fc)
checks.append([f.picked for f in ground.world.fruit] == before)   # reality untouched
fc2 = predict(twin, sync_report=ground.report(), rng_seed=7)
fc3 = predict(twin, sync_report=ground.report(), rng_seed=7)
checks.append(fc2['harvested'] == fc3['harvested'])               # reproducible
print('reality untouched + reproducible forecast: OK')
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
