# Lesson 8.4 — Course Close: From Mathematics to a Digital Twin

The closing verification: run the full twin-in-the-loop harvest once more and confirm the whole system completes. A clean 'All checks passed.' is the curriculum's last green light.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, harvest_row
from modules.module10.twin import (GroundTruth, DigitalTwin, snapshot,
    prevalidate, select_action, twin_in_the_loop)
mk = lambda: GreenhouseWorld.demo_row(n=5, seed=1)
F2 = mk().fruit[2]
OBST = {'F2': {'obstacle': ((float(F2.x), float(F2.y)), 0.05)}}
def effort(rep): return sum(p['n_attempts'] for p in rep['picks'])
checks = []
real = GroundTruth(mk())
twin = DigitalTwin(mk()); twin.sync(snapshot(real.world))
candidates = {'clear_path': None, 'ignore_obstacle': OBST}
# represent -> perceive -> reach -> move -> integrate -> twin, run end to end over the row
log = []
for _ in mk().fruit:
    turn = twin_in_the_loop(twin, snapshot(real.world), candidates)
    log.append(turn['chosen'])
final = harvest_row(mk(), max_attempts=3)
print('closing harvest:', sorted(final['harvested']), '| complete:', final['complete'])
print('loop turns:', len(log))
```

```python
checks.append(final['complete'] is True)
checks.append(len(final['harvested']) >= 4)
checks.append(len(log) == len(mk().fruit))
# the through-line ran: represent -> perceive -> reach -> move -> integrate -> twin
```

```python
assert all(checks), checks
print('All checks passed.')
```
