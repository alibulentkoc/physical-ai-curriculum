# Lesson 1.2 — Twin vs. Model vs. Simulation

A *simulation* runs the model forward on some world; a *twin* is bound to a **specific** real world and tracks it. Only the twin's state follows the specific real robot after a sync.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
from modules.module09.integration import harvest_row
checks = []
# SIMULATION: run the M9 model forward on a STANDALONE world (no real asset)
standalone = GreenhouseWorld.demo_row(n=6, seed=99)
sim = harvest_row(standalone)
print('simulation on a standalone world -> complete:', sim['complete'])
# TWIN: bound to a SPECIFIC real world; mirror it
real.q = np.array([0.3, 0.9])
twin = DigitalTwin(real); twin.sync(snapshot(real))
checks.append(twin.divergence(snapshot(real))['synced'])      # tracks THIS real world
# the standalone simulation does NOT track the specific real world
checks.append(not twin.divergence(snapshot(standalone))['synced'])
print('twin tracks the specific real world; standalone sim does not')
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
