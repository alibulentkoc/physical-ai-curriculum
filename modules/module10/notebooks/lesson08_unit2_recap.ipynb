The full mirror: sync -> synced; act without sync -> drift; re-sync -> recover; hidden offset -> residual gap. The twin mirrors but does not yet simulate.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
ground = GroundTruth(real)
twin = DigitalTwin(real)
twin.sync(ground.report()); checks.append(twin.divergence(ground.report())['synced'])   # synced
real.q = np.array([1.0, 0.1]); checks.append(not twin.divergence(snapshot(real))['synced'])  # drift
twin.sync(snapshot(real)); checks.append(twin.divergence(snapshot(real))['synced'])     # recover
g2 = GroundTruth(GreenhouseWorld.demo_row(n=6, seed=1), q_offset=np.array([0.04,0.04]))
t2 = DigitalTwin(g2.world); t2.sync(g2.report())
checks.append(t2.divergence(g2.true_report())['q_gap'] > 0.01)                            # residual gap
print('synced=%s drift=%s recover=%s residual_gap=%s' % tuple(checks))
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('INSTALLMENT A COMPLETE: the twin mirrors the real robot, live. Next: simulate.')
print('All checks passed.')
```
