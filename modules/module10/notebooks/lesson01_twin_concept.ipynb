# Lesson 1.1 — The Twin Concept: A Living Mirror

A digital twin is a **paired, live, mirroring** replica of a specific real system. Here we build one over the Module 9 robot and confirm it mirrors the real system's reported state after a sync.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
twin = DigitalTwin(real)
checks.append(twin.world is not real)          # the twin holds its OWN world
print('twin world is a separate instance:', twin.world is not real)
# move the real robot, then sync the twin to its reported state
real.q = np.array([0.2, 1.1])
twin.sync(snapshot(real))
d = twin.divergence(snapshot(real))
print('after sync -> divergence:', {k:(round(v,4) if isinstance(v,float) else v) for k,v in d.items()})
checks.append(d['synced'])                      # mirrors the reported state exactly
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
