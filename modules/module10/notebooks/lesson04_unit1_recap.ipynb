# Lesson 1.4 — Unit 1 Recap: The Twin Concept

Consolidation: the twin is a separate, paired replica that wraps the M9 system and can read its reported state.

```python
import numpy as np
from modules.module09.integration import GreenhouseWorld, run_pipeline
from modules.module10.twin import DigitalTwin, GroundTruth, snapshot, clone_layout
real = GreenhouseWorld.demo_row(n=6, seed=1)
checks = []
twin = DigitalTwin(real)
checks.append(twin.world is not real)                              # separate
checks.append([f.fid for f in twin.world.fruit] == [f.fid for f in real.fruit])  # reuses layout
rep = snapshot(real)
checks.append(all(k in rep for k in ('q','tool_xy','fruit','health','stage')))   # reads reported state
print('twin: separate =', checks[0], '| reuses layout =', checks[1], '| reads state =', checks[2])
assert all(checks), f'FAILED: {checks}'
print(f'{sum(checks)}/{len(checks)} checks passed.')
print('All checks passed.')
```
