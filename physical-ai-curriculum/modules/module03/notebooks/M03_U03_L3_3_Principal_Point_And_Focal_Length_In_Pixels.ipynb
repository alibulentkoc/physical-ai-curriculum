```python
import numpy as np
def make_K(fx,fy,cx,cy): return np.array([[fx,0,cx],[0,fy,cy],[0,0,1.0]])
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
K=make_K(805,802,318,244)  # a calibration result
print('optical-axis point -> principal point', np.round(project_K([0,0,1.0],K),1))
print('direction X/Z=0.2 -> u =', round(project_K([0.2,0,1.0],K)[0],1))
```

## Coding Exercise (§8)
Implement a simple 'is this K reasonable?' check (principal point near center, fx≈fy).

```python
# YOUR CODE HERE
import numpy as np
def k_reasonable(K,W,H,tol_c=0.1,tol_f=0.05):
    cx,cy=K[0,2],K[1,2]; fx,fy=K[0,0],K[1,1]
    near_center = abs(cx-W/2)<tol_c*W and abs(cy-H/2)<tol_c*H
    square = abs(fx-fy)/fx < tol_f
    return near_center and square
print('K reasonable:', k_reasonable(K,640,480))
bad=make_K(800,800,600,50); print('bad K reasonable:', k_reasonable(bad,640,480))
```
