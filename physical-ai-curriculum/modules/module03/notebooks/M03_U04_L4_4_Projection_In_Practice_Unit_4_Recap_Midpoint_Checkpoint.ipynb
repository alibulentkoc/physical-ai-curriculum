```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]]); W,H=640,480
def world_to_pixel(P_w,T,K):
    Pw=np.array([*P_w,1.0]); Pc=(T@Pw)[:3]; p=K@Pc; return p[:2]/p[2], Pc[2]>0
# full pipeline at a non-identity pose
T=np.eye(4); T[:3,3]=[-0.05,0.0,0.0]
uv,front=world_to_pixel([0.06,-0.03,0.3],T,K)
print('pixel', np.round(uv,1), 'in front:', bool(front))
```

## Coding Exercise (§8)
Confirm the pipeline against OpenCV (or NumPy fallback) and classify points.

```python
# YOUR CODE HERE
import numpy as np
def project_numpy(pts,R,t,K):
    pts=np.asarray(pts,float); Pc=(R@pts.T).T+t; p=(K@Pc.T).T; return p[:,:2]/p[:,2:3]
ref=project_numpy([[0.06,-0.03,0.3]],np.eye(3),np.zeros(3),K)
try:
    import cv2
    cv,_=cv2.projectPoints(np.array([[0.06,-0.03,0.3]]),np.zeros(3),np.zeros(3),K,np.zeros(5))
    print('matches OpenCV:', np.allclose(ref,cv.reshape(-1,2),atol=1e-6))
except Exception:
    print('OpenCV unavailable; NumPy pipeline result:', np.round(ref,1))
```
