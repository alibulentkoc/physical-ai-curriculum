```python
import numpy as np
K=np.array([[800,0,320],[0,800,240],[0,0,1.0]])
def deproject(u,v,Z,K): return np.array([(u-K[0,2])*Z/K[0,0],(v-K[1,2])*Z/K[1,1],Z])
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
P=deproject(480,160,0.3,K)
print('P_c (camera frame):',P,'  still NOT world frame (Unit 7 transforms it)')
```

## Coding Exercise (§8)
Deproject two pixels, verify round-trips, note camera-frame.

```python
# YOUR CODE HERE
import numpy as np
for (u,v,Z) in [(480,160,0.3),(320,240,0.5)]:
    Pc=deproject(u,v,Z,K); assert np.allclose(project_K(Pc,K),[u,v])
print('round-trips OK; points are in the camera frame.')
```
