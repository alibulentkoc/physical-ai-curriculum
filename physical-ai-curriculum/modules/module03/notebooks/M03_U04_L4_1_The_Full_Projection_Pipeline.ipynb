```python
import numpy as np
def make_K(fx,fy,cx,cy): return np.array([[fx,0,cx],[0,fy,cy],[0,0,1.0]])
def world_to_pixel(P_w, T_cam_world, K):
    Pw=np.array([*P_w,1.0]); Pc=(T_cam_world@Pw)[:3]
    p=K@Pc; return p[:2]/p[2]
K=make_K(800,800,320,240)
I=np.eye(4)
print('identity pose:', np.round(world_to_pixel([0.06,-0.03,0.3], I, K),1))
```
