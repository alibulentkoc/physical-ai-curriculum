### Lesson 5.3 — Undistortion

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def distort(xn,yn,k1=0,k2=0,k3=0,p1=0,p2=0):
    r2=xn*xn+yn*yn; rad=1+k1*r2+k2*r2**2+k3*r2**3
    xr,yr=xn*rad,yn*rad
    xt=2*p1*xn*yn+p2*(r2+2*xn*xn); yt=p1*(r2+2*yn*yn)+2*p2*xn*yn
    return xr+xt, yr+yt
def undistort(xd,yd,iters=10,**k):
    xn,yn=xd,yd
    for _ in range(iters):
        dx,dy=distort(xn,yn,**k)
        xn=xn+(xd-dx); yn=yn+(yd-dy)
    return xn,yn
# round trip
xd,yd=distort(0.35,0.0,k1=-0.2)
xn,yn=undistort(xd,yd,k1=-0.2)
print('recovered:', round(xn,4), round(yn,4), '(orig 0.35, 0.0)')
```
