```python
import numpy as np
def distort(xn,yn,k1=0,k2=0,k3=0,p1=0,p2=0):
    r2=xn*xn+yn*yn; rad=1+k1*r2+k2*r2**2+k3*r2**3
    return xn*rad+2*p1*xn*yn+p2*(r2+2*xn*xn), yn*rad+p1*(r2+2*yn*yn)+2*p2*xn*yn
def undistort(xd,yd,iters=15,**k):
    xn,yn=np.array(xd,float),np.array(yd,float)   # copy: never mutate inputs
    for _ in range(iters):
        dx,dy=distort(xn,yn,**k)
        xn=xn+(xd-dx); yn=yn+(yd-dy)
    return xn,yn
g=np.linspace(-0.4,0.4,9); X,Y=np.meshgrid(g,g)
Xd,Yd=distort(X,Y,k1=-0.25)
Xu,Yu=undistort(Xd,Yd,k1=-0.25)
print('grid round-trip max err:', f'{float(np.max(np.hypot(Xu-X,Yu-Y))):.2e}')
```

## Coding Exercise (§8)
Distort then undistort a whole grid; confirm round-trip recovers originals; center unaffected.

```python
# YOUR CODE HERE
import numpy as np
assert np.allclose(np.stack([Xu,Yu]),np.stack([X,Y]),atol=1e-4)
cx,cy=undistort(np.array(0.0),np.array(0.0),k1=-0.25)
assert np.allclose([float(cx),float(cy)],[0,0])   # center unaffected
print('round-trip recovers the grid; center unaffected.')
```
