```python
import numpy as np
f=800.0
def project(P): X,Y,Z=P; return np.array([f*X/Z,f*Y/Z])
def backproject(pix,Z): return np.array([pix[0]*Z/f, pix[1]*Z/f, Z])
P=np.array([0.06,-0.03,0.3])
pix=project(P)
print('forward: point', P, '-> pixel', np.round(pix,1))
print('same ray, Z=0.6 -> pixel', np.round(project(np.array([0.12,-0.06,0.6])),1), '(same)')
print('inverse with depth 0.3 ->', np.round(backproject(pix,0.3),3))
```

## Coding Exercise (§8)
Round trip: project, then back-project with the correct depth to recover the point.

```python
# YOUR CODE HERE
import numpy as np
assert np.allclose(backproject(project(P), P[2]), P)
print('round trip with correct depth recovers the point.')
```
