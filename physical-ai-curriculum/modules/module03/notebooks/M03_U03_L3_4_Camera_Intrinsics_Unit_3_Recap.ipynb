```python
import numpy as np
def make_K(fx,fy,cx,cy): return np.array([[fx,0,cx],[0,fy,cy],[0,0,1.0]])
def project_K(P,K): p=K@np.asarray(P,float); return p[:2]/p[2]
K=make_K(800,800,320,240)
# focal length scales offset from principal point
off1=project_K([0.2,0,1.0],make_K(800,800,320,240))[0]-320
off2=project_K([0.2,0,1.0],make_K(1600,1600,320,240))[0]-320
print('offset doubles with f:', np.isclose(off2,2*off1))
```

## Coding Exercise (§8)
Project several points; confirm axis point → principal point and offset ∝ f.

```python
# YOUR CODE HERE
import numpy as np
pts=[[0,0,0.5],[0.06,-0.03,0.3],[0.1,0.1,0.4]]
for P in pts: print(P,'->',np.round(project_K(P,K),1))
assert np.allclose(project_K([0,0,0.5],K),[320,240])
```
