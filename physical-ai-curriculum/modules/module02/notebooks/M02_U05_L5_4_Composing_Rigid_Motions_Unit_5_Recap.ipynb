```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
A=se3(Rz(90),[0,0,0]); B=se3(Rz(0),[2,0,0])
print('order matters:', not np.allclose(A@B, B@A))
print('inverse reverses path:', np.allclose(inv(B@A), inv(A)@inv(B)))
```
