```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
# detection carries an orientation too (a full pose), not just a point
T_cam_obj   = se3(Rz(20),[0,0,0.3])        # tomato pose in camera frame
T_arm_cam   = se3(np.eye(3),[0,0,0.1])     # extrinsics
T_world_arm = se3(Rz(0),[1.0,0.5,0.0])     # arm pose
T_world_obj = T_world_arm @ T_arm_cam @ T_cam_obj
print('tomato world pose position:', np.round(T_world_obj[:3,3],3))
```

## Coding Exercise (§8)
Compose the full chain to a world pose, then invert to confirm the round trip.

```python
# YOUR CODE HERE
import numpy as np
T_world_cam=T_world_arm@T_arm_cam
recovered = inv(T_world_cam)@T_world_obj   # tomato back in camera frame
print('round trip = original camera pose:', np.allclose(recovered,T_cam_obj))
```
