```python
import numpy as np
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
def Rx(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[1,0,0],[0,c,-s],[0,s,c]])
def Ry(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,0,s],[0,1,0],[-s,0,c]])
print('Rz(90) @ (1,0,0) =', np.round(Rz(90)@np.array([1,0,0.]),3), '(z unchanged)')
print('Rx(90) @ (0,1,0) =', np.round(Rx(90)@np.array([0,1,0.]),3), '(tipped up)')
```

## Coding Exercise (§8)
Rotate points and confirm length is preserved and the on-axis coordinate is unchanged.

```python
# YOUR CODE HERE
import numpy as np
p=np.array([1,2,3.])
for name,R in [('Rz',Rz(37)),('Rx',Rx(37)),('Ry',Ry(37))]:
    print(name,'length preserved:', np.isclose(np.linalg.norm(R@p),np.linalg.norm(p)))
```
