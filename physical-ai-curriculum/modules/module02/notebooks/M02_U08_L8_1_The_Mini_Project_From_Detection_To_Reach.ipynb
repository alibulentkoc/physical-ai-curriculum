**The problem.** Given a detected tomato pose in the camera frame, the camera's extrinsics (its pose on the arm), and the arm's pose in the world, find the tomato's pose in the **world** frame:

$$T_{world\leftarrow tomato} = T_{world\leftarrow arm}\;T_{arm\leftarrow cam}\;T_{cam\leftarrow tomato}.$$

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
I=np.eye(3)
# scaffold: the three given poses (placeholder values, filled in 8.2)
T_cam_tomato = se3(I,[0,0,0.3])     # detection in camera frame
T_arm_cam    = se3(I,[0,0,0.1])     # extrinsics
T_world_arm  = se3(I,[1.0,0.5,0])   # arm pose in world
# composition that produces the answer
T_world_tomato = T_world_arm @ T_arm_cam @ T_cam_tomato
print('scaffold composed; world position =', np.round(T_world_tomato[:3,3],3))
```
