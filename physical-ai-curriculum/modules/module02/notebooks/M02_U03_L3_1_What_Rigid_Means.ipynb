## Unit 3 — SE(2) Transformations
### Lesson 3.1 — What "Rigid" Means

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se2(theta_deg,tx,ty):
    a=np.radians(theta_deg); c,s=np.cos(a),np.sin(a)
    return np.array([[c,-s,tx],[s,c,ty],[0,0,1.0]])
def scale(sx,sy): return np.array([[sx,0,0],[0,sy,0],[0,0,1.0]])
def apply(T,pts):  # pts: 2xN -> homogeneous -> 2xN
    h=np.vstack([pts,np.ones(pts.shape[1])]); return (T@h)[:2]
def pairwise(pts):
    import itertools; return [round(np.linalg.norm(pts[:,i]-pts[:,j]),4) for i,j in itertools.combinations(range(pts.shape[1]),2)]
sq=np.array([[0,1,1,0],[0,0,1,1.0]])
rigid=apply(se2(50,1.3,0.4),sq)
print('rigid: distances preserved:', pairwise(sq)==pairwise(rigid))
nonrigid=apply(scale(2,1),sq)
print('scaling: distances preserved:', pairwise(sq)==pairwise(nonrigid))
```
