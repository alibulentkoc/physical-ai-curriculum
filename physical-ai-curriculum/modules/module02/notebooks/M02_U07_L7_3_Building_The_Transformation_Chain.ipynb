## Unit 7 — Camera-to-Robot Transformations
### Lesson 7.3 — Building the Transformation Chain

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
T_arm_cam   = se3(np.eye(3),[0,0,0.1])    # extrinsics
T_world_arm = se3(np.eye(3),[1.0,0.5,0.0]) # arm pose
T_world_cam = T_world_arm @ T_arm_cam      # composed chain
p_cam=np.array([0,0,0.3,1.0])
print('camera->world product:\n', np.round(T_world_cam,3))
print('detection in world:', np.round((T_world_cam@p_cam)[:3],3))
```

## Coding Exercise (§8)
Invert the chain to bring a world goal into the camera frame; confirm a round trip.

```python
# YOUR CODE HERE
import numpy as np
p_world=T_world_cam@p_cam
back=inv(T_world_cam)@p_world
print('round trip back to camera frame:', np.round(back[:3],3))
print('matches original p_cam:', np.allclose(back,p_cam))
```
