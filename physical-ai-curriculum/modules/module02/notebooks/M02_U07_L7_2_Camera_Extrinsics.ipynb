### Lesson 7.2 — Camera Extrinsics

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
# extrinsics: camera 0.1 m out along wrist z, aligned with wrist
T_arm_cam=se3(np.eye(3),[0,0,0.1])
p_cam=np.array([0,0,0.3,1.0])   # 0.3 m in front of camera
p_arm=T_arm_cam@p_cam
print('extrinsics T(arm<-cam):\n', np.round(T_arm_cam,3))
print('detection in arm frame:', np.round(p_arm[:3],3))
```

Change the extrinsics (add a tilt) and show the arm-frame result changes.

```python
# YOUR CODE HERE
import numpy as np
def Rx(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[1,0,0],[0,c,-s],[0,s,c]])
T_tilt=se3(Rx(30),[0,0,0.1])
print('aligned :', np.round((T_arm_cam@p_cam)[:3],3))
print('tilted  :', np.round((T_tilt@p_cam)[:3],3))
```
