```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti

def detection_to_world_target(T_world_arm, T_arm_cam, T_cam_tomato, check=True):
    """Full perception-to-pose pipeline with built-in verification."""
    T_world_cam = T_world_arm @ T_arm_cam
    T_world_tomato = T_world_cam @ T_cam_tomato
    if check:
        assert np.allclose(inv(T_world_cam)@T_world_tomato, T_cam_tomato), 'round-trip failed'
        d_cam=np.linalg.norm(T_cam_tomato[:3,3]); d_world=np.linalg.norm(T_world_tomato[:3,3]-T_world_cam[:3,3])
        assert np.isclose(d_cam,d_world), 'distance not preserved'
        assert np.allclose(T_world_tomato[3],[0,0,0,1])
    return T_world_tomato
```
