```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def Rz(d):
    a=np.radians(d); c,s=np.cos(a),np.sin(a); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
def inv(T):
    R=T[:3,:3]; t=T[:3,3]; Ti=np.eye(4); Ti[:3,:3]=R.T; Ti[:3,3]=-R.T@t; return Ti
# edges
T_arm_cam   = se3(Rz(0),[0.0,0.0,-0.2])   # camera -> arm
T_world_arm = se3(Rz(90),[1.0,0.5,0.0])   # arm -> world
# path camera -> world
T_world_cam = T_world_arm @ T_arm_cam
p_cam=np.array([0.3,0,0,1.0])
print('tomato in world:', np.round((T_world_cam@p_cam)[:3],3))
```

## Coding Exercise (§8)
Invert the path to get world->camera; confirm world->camera ∘ camera->world is identity.

```python
# YOUR CODE HERE
import numpy as np
T_cam_world = inv(T_world_cam)
print('round trip = identity:', np.allclose(T_cam_world @ T_world_cam, np.eye(4)))
# also equals inverse-of-each in reverse order
print('matches reversed inverses:', np.allclose(T_cam_world, inv(T_arm_cam)@inv(T_world_arm)))
```
