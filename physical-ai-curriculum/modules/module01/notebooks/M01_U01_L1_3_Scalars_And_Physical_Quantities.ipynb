```python
# A scalar reading: value + unit. Direction is absent.
quantities = [
    {'name':'air temperature', 'value':28.0, 'unit':'C',   'needs_direction':False},
    {'name':'basket mass',     'value':1.5,  'unit':'kg',  'needs_direction':False},
    {'name':'gripper speed',   'value':0.2,  'unit':'m/s', 'needs_direction':False},
    {'name':'gripper velocity','value':0.2,  'unit':'m/s', 'needs_direction':True},
    {'name':'tomato position', 'value':None, 'unit':'m',   'needs_direction':True},
]
for q in quantities:
    kind = 'needs direction (vector)' if q['needs_direction'] else 'scalar'
    print(f"{q['name']:18s} -> {kind}")
```
