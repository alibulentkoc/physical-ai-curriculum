```python
import numpy as np
A=np.array([[0,-1],[1,0]])     # rotate 90 (first action)
B=np.array([[2,0],[0,2]])      # scale x2 (second action)
p=np.array([1.0,0.0])
seq = B@(A@p)                  # do A then B
prod = (B@A)@p                 # apply product BA
print('B(Ap) =', seq, '  (BA)p =', prod, '  equal:', np.allclose(seq,prod))
print('order matters: AB == BA?', np.allclose(A@B, B@A))
```
