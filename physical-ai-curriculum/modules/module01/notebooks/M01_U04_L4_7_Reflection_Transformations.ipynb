### Lesson 4.7 — Reflection Transformations

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
Fx = np.array([[1,0],[0,-1]])   # reflect across x-axis (flip y)
Fy = np.array([[-1,0],[0,1]])   # reflect across y-axis (flip x)
p = np.array([2.0,3.0])
print('reflect across x-axis:', Fx@p)
print('reflect across y-axis:', Fy@p)
print('det(Fx) =', np.linalg.det(Fx), ' (−1 = orientation flip)')
print('reflect twice = identity:', np.allclose(Fx@Fx, np.eye(2)))
```

## Coding Exercise (§8)
Reflect a shape across the y-axis; confirm distances from the origin are preserved and det = −1.

```python
# YOUR CODE HERE
import numpy as np
pts=np.array([[1,0],[2,1],[1,2]]).T
refl=Fy@pts
print('distances preserved:', np.allclose(np.linalg.norm(pts,axis=0), np.linalg.norm(refl,axis=0)))
print('det(Fy) =', np.linalg.det(Fy))
```
