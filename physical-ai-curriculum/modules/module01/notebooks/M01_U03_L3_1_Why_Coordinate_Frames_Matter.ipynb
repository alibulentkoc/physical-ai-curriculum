```python
import numpy as np
# A fixed tomato in the WORLD frame (origin = greenhouse corner):
tomato_world = np.array([2.0, 1.5])
# The robot, also measured in the world frame (axes aligned, no rotation yet):
robot_world = np.array([1.4, 1.3])
# Same tomato, described from the ROBOT's origin = world point minus robot origin:
tomato_robot = tomato_world - robot_world
print('tomato in WORLD frame:', tomato_world)
print('tomato in ROBOT frame:', tomato_robot)
print('Same tomato. Different coordinates. The point did not move.')
```
