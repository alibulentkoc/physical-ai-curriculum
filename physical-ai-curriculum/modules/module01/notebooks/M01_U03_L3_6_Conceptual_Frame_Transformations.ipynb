```python
import numpy as np
def convert_A_to_B(point, offset, angle_deg):
    """Re-describe a point from frame A in frame B (offset + rotation), no matrices."""
    # 1) shift: position relative to B's origin
    shifted = np.array(point, float) - np.array(offset, float)
    # 2) rotate by -angle to read along B's turned axes
    a = np.radians(angle_deg)
    x, y = shifted
    xb =  x*np.cos(a) + y*np.sin(a)
    yb = -x*np.sin(a) + y*np.cos(a)
    return np.array([xb, yb])

P_A = [2.0, 1.5]
print('offset only (theta=0):', convert_A_to_B(P_A, [1.0,0.5], 0).round(3))   # matches 3.5 subtraction
print('offset + 90 deg rot  :', convert_A_to_B(P_A, [1.0,0.5], 90).round(3))
```

## Coding Exercise (§8)
Confirm the offset-only case equals plain subtraction (Lesson 3.5), and that the point's distance from B's origin is unchanged by rotation.

```python
# YOUR CODE HERE
offset=[1.0,0.5]
sub_only = np.array(P_A)-np.array(offset)
print('matches 3.5 subtraction:', np.allclose(convert_A_to_B(P_A,offset,0), sub_only))
import numpy as np
print('distance preserved under rotation:', np.allclose(np.linalg.norm(convert_A_to_B(P_A,offset,0)), np.linalg.norm(convert_A_to_B(P_A,offset,37))))
```
