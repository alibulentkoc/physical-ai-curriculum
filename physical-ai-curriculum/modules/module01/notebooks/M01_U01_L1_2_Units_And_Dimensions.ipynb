```python
def cm_to_m(v):  return v / 100.0
def mm_to_m(v):  return v / 1000.0
def deg_to_rad(v): return v * 3.141592653589793 / 180.0

print(cm_to_m(30))     # 0.3  (length)
print(mm_to_m(1200))   # 1.2  (length)
print(deg_to_rad(90))  # ~1.5708 (angle)
```

## Dimension check
A tiny helper: dimensions are tracked as exponents of (L, M, T). Two quantities can be added only if their dimension tuples match.

```python
def can_add(dim_a, dim_b):
    return dim_a == dim_b

L=(1,0,0); T=(0,0,1); velocity=(1,0,-1)
print('length + length:', can_add(L, L))           # True
print('length + time:  ', can_add(L, T))           # False
print('d/t is a velocity:', (1,0,0-1)==velocity)   # True
```
