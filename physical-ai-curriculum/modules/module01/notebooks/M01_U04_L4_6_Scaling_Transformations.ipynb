```python
import numpy as np
def S(sx,sy): return np.array([[sx,0],[0,sy]])
p=np.array([2.0,3.0])
print('scale (2,0.5):', S(2,0.5)@p)
print('uniform scale 3 on (2,3):', S(3,3)@p)
```

## Coding Exercise (§8)
Scale a unit square and confirm its area scales by sx*sy.

```python
# YOUR CODE HERE
import numpy as np
sx,sy=3,2
square=np.array([[0,0],[1,0],[1,1],[0,1]]).T
scaled=S(sx,sy)@square
width=scaled[0].max()-scaled[0].min(); height=scaled[1].max()-scaled[1].min()
print('new area:', width*height, ' expected sx*sy:', sx*sy)
```
