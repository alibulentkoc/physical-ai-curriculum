```python
import numpy as np
a = np.array([1.0, 0.0, 0.0])
b = np.array([0.0, 1.0, 0.0])
n = np.cross(a, b)
print('a x b =', n, ' (perpendicular to both)')
print('dot with a:', np.dot(n,a), ' dot with b:', np.dot(n,b), ' (both 0)')
print('area of parallelogram = |a x b| =', np.linalg.norm(n))
```
