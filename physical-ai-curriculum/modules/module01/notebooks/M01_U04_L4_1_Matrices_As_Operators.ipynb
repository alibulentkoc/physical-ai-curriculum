```python
import numpy as np
p = np.array([2.0, 0.0])
M_turn = np.array([[0,-1],[1,0]])      # 90 deg turn
M_grow = np.array([[2,0],[0,2]])       # double size
print('p          :', p)
print('M_turn @ p :', M_turn @ p, '(east -> north: a 90 deg turn)')
print('M_grow @ p :', M_grow @ p, '(twice as far: scaling)')
print('columns of M_turn = where e1,e2 land:', M_turn[:,0], M_turn[:,1])
```

## Coding Exercise (§8)
Apply M to a whole shape (list of points) at once and say what it did.

```python
# YOUR CODE HERE
shape = np.array([[0,0],[1,0],[1,1],[0,1]]).T   # unit square columns
out = M_grow @ shape
print('scaled square columns:\n', out)
print('it doubled the square (action = uniform scale by 2).')
```
