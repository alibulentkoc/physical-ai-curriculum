```python
import numpy as np
# what the camera sees, in the CAMERA frame:
tomato_cam = np.array([0.30, -0.10])
# camera mounted on the robot: offset of camera origin in the ROBOT frame
cam_in_robot = np.array([0.05, 0.25])
# robot base pose in the WORLD frame
robot_in_world = np.array([1.65, 1.35])

tomato_robot = tomato_cam + cam_in_robot       # camera -> robot
tomato_world = tomato_robot + robot_in_world    # robot  -> world
print('camera:', tomato_cam)
print('robot :', tomato_robot)
print('world :', tomato_world)
```

## Coding Exercise (§8)
Wrap the chain in a function camera_to_world(p_cam) and test it on a second detection (0.10, 0.05).

```python
# YOUR CODE HERE
def camera_to_world(p_cam):
    return p_cam + cam_in_robot + robot_in_world
print('detection 2 in world:', camera_to_world(np.array([0.10, 0.05])))
```
