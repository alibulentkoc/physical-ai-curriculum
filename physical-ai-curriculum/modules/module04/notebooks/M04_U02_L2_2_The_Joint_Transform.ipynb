### Lesson 2.2 — The Joint Transform

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def rot2(theta):
    c,s=np.cos(theta),np.sin(theta); return np.array([[c,-s,0],[s,c,0],[0,0,1.0]])
def trans2(dx,dy):
    return np.array([[1,0,dx],[0,1,dy],[0,0,1.0]])
def joint_transform(kind,q,L):
    G=trans2(L,0)              # fixed link along x
    if kind=='revolute': return rot2(q)@G
    else:                return trans2(q,0)@G   # prismatic slides along x
T=joint_transform('revolute',np.radians(60),0.4)
print('revolute 60deg, L=0.4 -> translation', np.round(T[:2,2],3))
```
