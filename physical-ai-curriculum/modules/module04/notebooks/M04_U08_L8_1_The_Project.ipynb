```python
import numpy as np
# Capstone inputs: 3-DOF DH table, camera mount, a perceived fruit.
arm3=[(0.1,0,np.radians(90),'r'),(0,0.4,0,'r'),(0,0.3,0,'r')]   # (d,a,alpha,kind)
def se3(R,t):
    T=np.eye(4);T[:3,:3]=R;T[:3,3]=t;return T
T_base_cam=se3(np.eye(3),[0.1,0,0.5])
P_cam=np.array([0.05,0,0.40])
def apply(M,P): return (M@np.r_[np.asarray(P,float),1.0])[:3]
P_base=apply(T_base_cam,P_cam)
print('inputs ready. fruit in base frame:',np.round(P_base,3))
```
