```python
import numpy as np
def summarize_one_joint(theta,L):
    return {'dof':1,'config':('theta1',),'pos':np.array([L*np.cos(theta),L*np.sin(theta)])}
s=summarize_one_joint(np.radians(60),0.5)
print('DOF',s['dof'],'config',s['config'],'pos',np.round(s['pos'],3))
```
