### Lesson 3.1 — Adding a Second Joint

*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
def fk_two_link(t1,t2,L1,L2):
    x=L1*np.cos(t1)+L2*np.cos(t1+t2)
    y=L1*np.sin(t1)+L2*np.sin(t1+t2)
    return np.array([x,y]), t1+t2   # position, orientation
L1,L2=0.4,0.3
p,ori=fk_two_link(np.radians(30),np.radians(60),L1,L2)
print('gripper',np.round(p,3),'orientation(deg)',round(np.degrees(ori)))
```
