```python
import numpy as np
def pose_link(theta,L):
    c,s=np.cos(theta),np.sin(theta); return np.array([[c,-s,L*c],[s,c,L*s],[0,0,1.0]])
def fk_chain(joints):
    T=np.eye(3)
    for theta,L in joints: T=T@pose_link(theta,L)
    return T[:2,2], np.arctan2(T[1,0],T[0,0])
def fk_planar(joints):
    phi=0.0;x=0.0;y=0.0
    for theta,L in joints:
        phi+=theta; x+=L*np.cos(phi); y+=L*np.sin(phi)
    return np.array([x,y]),phi
js=[(np.radians(30),0.4),(np.radians(60),0.3),(np.radians(-30),0.2)]
print('chain',np.round(fk_chain(js)[0],3),' planar',np.round(fk_planar(js)[0],3))
```

## Coding Exercise (§8)
Confirm both methods agree on the 3-link example, position and orientation.

```python
# YOUR CODE HERE
import numpy as np
pc,oc=fk_chain(js); pp,op=fk_planar(js)
assert np.allclose(pc,pp,atol=1e-9)
assert np.isclose(((oc-op+np.pi)%(2*np.pi))-np.pi,0,atol=1e-9)
print('product and closed form agree (position + orientation).')
```
