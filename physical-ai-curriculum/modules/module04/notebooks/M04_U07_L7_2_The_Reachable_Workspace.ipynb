*Physical AI Curriculum · hands-on notebook. Run **Kernel → Restart & Run All**.*

```python
import numpy as np
import matplotlib; matplotlib.use('Agg'); import matplotlib.pyplot as plt
L1,L2=0.4,0.3
def fk2(t1,t2):
    return np.array([L1*np.cos(t1)+L2*np.cos(t1+t2), L1*np.sin(t1)+L2*np.sin(t1+t2)])
rng=np.random.default_rng(0)
Q=rng.uniform(0,2*np.pi,size=(4000,2))
P=np.array([fk2(*q) for q in Q])
r=np.linalg.norm(P,axis=1)
print('reach range:',round(r.min(),3),'to',round(r.max(),3))
```

## Coding Exercise (§8)
Plot the workspace and verify all points lie in the annulus [|L1-L2|, L1+L2].

```python
# YOUR CODE HERE
import numpy as np
fig,ax=plt.subplots(); ax.scatter(P[:,0],P[:,1],s=1); ax.set_aspect('equal'); fig.savefig('/tmp/ws.png')
assert r.max()<=L1+L2+1e-9 and r.min()>=abs(L1-L2)-1e-9
# reachability test
def reachable(d): return abs(L1-L2)-1e-9 <= d <= L1+L2+1e-9
assert reachable(0.5) and not reachable(0.05) and not reachable(0.8)
print('workspace annulus verified; reachability test works.')
```
