```python
import numpy as np
def se3(R,t):
    T=np.eye(4); T[:3,:3]=R; T[:3,3]=t; return T
def apply(T,P):
    Ph=np.r_[np.asarray(P,float),1.0]; return (T@Ph)[:3]
# camera mounted on base: translation only
T_base_cam=se3(np.eye(3),[0.1,0,0.5])
P_cam=np.array([0.05,0,0.40])
P_base=apply(T_base_cam,P_cam)
print('fruit in base frame',np.round(P_base,3))
```

## Coding Exercise (§8)
Place the fruit relative to the gripper and report the base-frame move vector.

```python
# YOUR CODE HERE
import numpy as np
# a current gripper pose T_0^n (example): identity rotation, position (0.15,0,0.60)
T0n=se3(np.eye(3),[0.15,0,0.60])
P_grip=apply(np.linalg.inv(T0n),P_base)
move=P_base-T0n[:3,3]
assert np.allclose(P_base,[0.15,0,0.90])
assert np.allclose(move,[0,0,0.30])
print('fruit rel. gripper',np.round(P_grip,3),' move vector',np.round(move,3))
```
